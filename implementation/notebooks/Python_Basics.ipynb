## 1. Variables and Types

Python infers types at assignment — no `int x = 5;` declarations. This matters because AI frameworks pass data between functions where types change silently.

```python
# WHAT: Declare variables and check their types.
# WHY: Every ML pipeline starts with data that could be int, float, or str.
# Knowing the type tells you what operations are safe.

model_accuracy = 0.94          # float — most ML metrics are floats
epoch_count = 50               # int — you can't run 50.5 epochs
model_name = "bert-base"       # str — model registry lookups use strings
is_trained = True              # bool — feature flags, pipeline gates

print(f"model_accuracy: {model_accuracy} -> {type(model_accuracy)}")
print(f"epoch_count:    {epoch_count} -> {type(epoch_count)}")
print(f"model_name:     {model_name} -> {type(model_name)}")
print(f"is_trained:     {is_trained} -> {type(is_trained)}")

# You Should See:
# model_accuracy: 0.94 -> <class 'float'>
# epoch_count:    50 -> <class 'int'>
# model_name:     bert-base -> <class 'str'>
# is_trained:     True -> <class 'bool'>
```

## 2. Strings and f-Strings

f-strings are Python's string interpolation — the equivalent of Java's `String.format()` or JavaScript's template literals. You will use them in every log message and report.

```python
# WHAT: Format strings with embedded expressions.
# WHY: Logging and reporting are 30% of production pipeline code.

pipeline_name = "daily_etl"
rows_processed = 1_250_000   # underscores are legal separators in Python
duration_sec = 47.3

# f-string with formatting: comma separator, 1 decimal place
log_msg = f"Pipeline '{pipeline_name}' processed {rows_processed:,} rows in {duration_sec:.1f}s"
print(log_msg)

# You Should See:
# Pipeline 'daily_etl' processed 1,250,000 rows in 47.3s
```

```python
# WHAT: Clean raw log lines with string methods.
# WHY: Raw data from logs and CSVs arrives dirty — these four methods
# handle 90% of string cleanup.

raw_log = "  ERROR | 2026-04-04 | model_server | OOM on batch 42  \n"

cleaned = raw_log.strip()                  # remove leading/trailing whitespace
parts = cleaned.split(" | ")               # split into fields
level = parts[0].lower()                   # normalize to lowercase
fixed = cleaned.replace("OOM", "Out of Memory")  # expand abbreviation

print(f"stripped:  '{cleaned}'")
print(f"split:     {parts}")
print(f"lowered:   '{level}'")
print(f"replaced:  '{fixed}'")

# You Should See:
# stripped:  'ERROR | 2026-04-04 | model_server | OOM on batch 42'
# split:     ['ERROR', '2026-04-04', 'model_server', 'OOM on batch 42']
# lowered:   'error'
# replaced:  'ERROR | 2026-04-04 | model_server | Out of Memory on batch 42'
```

```python
# WHAT: Extract substrings with slicing.
# WHY: Timestamps and IDs often need substring extraction.
# Python slicing is [start:stop:step] — stop is exclusive.

timestamp = "2026-04-04T14:30:00Z"

date_part = timestamp[:10]        # first 10 chars
time_part = timestamp[11:19]      # chars 11-18
year = timestamp[:4]              # first 4 chars

print(f"Full:  {timestamp}")
print(f"Date:  {date_part}")
print(f"Time:  {time_part}")
print(f"Year:  {year}")

# You Should See:
# Full:  2026-04-04T14:30:00Z
# Date:  2026-04-04
# Time:  14:30:00
# Year:  2026
```

## 3. Operators

Arithmetic, comparison, and logical operators work like most languages — except `**` for power and `//` for integer division.

```python
# WHAT: Arithmetic operators for metrics and batch processing.
# WHY: Threshold checks and batch sizing drive every monitoring pipeline.

# Computing a metric
total_predictions = 10_000
correct = 9_400
accuracy = correct / total_predictions
print(f"Accuracy: {accuracy}")

# Integer division and modulo — batch processing
total_rows = 10_500
batch_size = 1_000
full_batches = total_rows // batch_size   # 10
remainder = total_rows % batch_size       # 500
print(f"Full batches: {full_batches}, remainder: {remainder}")

# Power — exponential backoff for retries
retry_attempt = 3
wait_seconds = 2 ** retry_attempt  # 8 seconds
print(f"Retry wait: {wait_seconds}s")

# You Should See:
# Accuracy: 0.94
# Full batches: 10, remainder: 500
# Retry wait: 8s
```

```python
# WHAT: Comparison and logical operators for pipeline gates.
# WHY: Pipeline gates decide whether to promote data or alert on-call.

accuracy = 0.94
latency_ms = 250
is_production = True

meets_sla = accuracy >= 0.90 and latency_ms < 500
needs_alert = accuracy < 0.85 or latency_ms > 1000
is_canary = is_production and not meets_sla

print(f"Meets SLA:   {meets_sla}")
print(f"Needs alert: {needs_alert}")
print(f"Is canary:   {is_canary}")

# You Should See:
# Meets SLA:   True
# Needs alert: False
# Is canary:   False
```

## 4. Type Conversion

Data arrives as strings from CSVs, APIs, and command-line arguments. You must convert explicitly.

```python
# WHAT: Convert string data to numeric types.
# WHY: CSV readers return strings. API responses return strings.
# If you add "5" + "3" you get "53", not 8.

# Simulating values read from a CSV row
raw_duration = "47.3"
raw_count = "1250000"
raw_flag = "1"

duration = float(raw_duration)
count = int(raw_count)
flag = bool(int(raw_flag))       # "1" -> 1 -> True

print(f"duration: {duration} ({type(duration).__name__})")
print(f"count:    {count} ({type(count).__name__})")
print(f"flag:     {flag} ({type(flag).__name__})")

# Common mistake: bool("False") is True! Non-empty strings are truthy.
print(f"\nbool('False') = {bool('False')}  # Surprise! Non-empty string = True")
print(f"bool('')      = {bool('')}   # Only empty string is False")

# You Should See:
# duration: 47.3 (float)
# count:    1250000 (int)
# flag:     True (bool)
# bool('False') = True  # Surprise! Non-empty string = True
# bool('')      = False   # Only empty string is False
```

## 5. None — Python's Null

None is Python's `null`/`nil`. You will see it everywhere: missing config values, uninitialized model weights, API fields that were not returned.

```python
# WHAT: Check for None and provide defaults.
# WHY: Null handling is the number one source of data pipeline bugs.
# In Python, use 'is None' — never '== None'.

model_checkpoint = None   # model hasn't been trained yet

# Correct null check — identity comparison
if model_checkpoint is None:
    print("No checkpoint found — training from scratch")

# Common pattern: default value if None
batch_size = None
effective_batch = batch_size if batch_size is not None else 32
print(f"Using batch size: {effective_batch}")

# Common mistake: 'or' treats 0 as falsy
batch_size = 0
bad_default = batch_size or 32
print(f"\n0 or 32 = {bad_default}  # Bug! 0 is falsy, so 'or' skips it")

# You Should See:
# No checkpoint found — training from scratch
# Using batch size: 32
# 0 or 32 = 32  # Bug! 0 is falsy, so 'or' skips it
```

## 6. Boolean Logic — Truthy and Falsy

In Python, these are all `False`: `0`, `0.0`, `""`, `[]`, `{}`, `None`, `False`. Everything else is `True`. This is how `if data:` works.

```python
# WHAT: Use truthy/falsy checks for defensive pipeline code.
# WHY: 'if results:' is idiomatic Python for 'if results is not empty'.

# Simulate checking pipeline outputs before writing
empty_results = []
valid_results = [{"id": 1, "score": 0.95}]

if not empty_results:
    print("No results to write — skipping output")

if valid_results:
    print(f"Writing {len(valid_results)} results to output")

# Full truth table of falsy values
print("\nFalsy values in Python:")
falsy_values = [0, 0.0, "", [], {}, None, False]
for val in falsy_values:
    print(f"  bool({str(val):>8}) = {bool(val)}")

# You Should See:
# No results to write — skipping output
# Writing 1 results to output
# Falsy values in Python:
#   bool(       0) = False
#   bool(     0.0) = False
#   ...all False
```

## 7. Control Flow

if/elif/else and for/while — the same as every language, but with Python's indentation-based blocks.

```python
# WHAT: Classify model performance with if/elif/else.
# WHY: Routing logic in data pipelines — classify, filter, transform.

accuracy = 0.87

if accuracy >= 0.95:
    status = "excellent"
elif accuracy >= 0.90:
    status = "good"
elif accuracy >= 0.80:
    status = "acceptable"
else:
    status = "needs_retraining"

print(f"Accuracy {accuracy} -> status: {status}")

# You Should See:
# Accuracy 0.87 -> status: acceptable
```

```python
# WHAT: Iterate over predictions with enumerate() and ternary expressions.
# WHY: Processing batches of results is constant in DE and ML work.

predictions = [0.92, 0.45, 0.88, 0.31, 0.97]
threshold = 0.5

for i, score in enumerate(predictions):
    label = "positive" if score >= threshold else "negative"
    print(f"  Prediction {i}: {score:.2f} -> {label}")

# Count positives with a generator expression
positive_count = sum(1 for s in predictions if s >= threshold)
print(f"\nTotal positive: {positive_count}/{len(predictions)}")

# You Should See:
#   Prediction 0: 0.92 -> positive
#   Prediction 1: 0.45 -> negative
#   Prediction 2: 0.88 -> positive
#   Prediction 3: 0.31 -> negative
#   Prediction 4: 0.97 -> positive
# Total positive: 3/5
```

```python
# WHAT: While loop with retry logic.
# WHY: Retries with backoff are a production pattern for API calls
# and database connections.

import random
random.seed(42)  # reproducible for the notebook

max_retries = 5
attempt = 0

while attempt < max_retries:
    attempt += 1
    success = random.random() > 0.6  # 40% chance of failure
    wait = 2 ** attempt
    
    if success:
        print(f"  Attempt {attempt}: Success!")
        break
    else:
        print(f"  Attempt {attempt}: Failed. Retry in {wait}s...")
else:
    # 'else' on a while loop runs if the loop finished WITHOUT break
    print("  All retries exhausted!")

# You Should See:
# A series of attempts, stopping on success or exhausting retries
```
