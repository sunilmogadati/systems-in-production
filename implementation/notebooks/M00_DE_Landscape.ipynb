## 1. What is Data Engineering?

**One-line answer:** Data engineers build and maintain the systems that collect, store, transform, and deliver data so that other people can use it.

That's it. Every data engineer job, at every company, boils down to that sentence.

> **Analogy:** If a restaurant is a company, the chef is the data scientist (makes the product), the server is the analyst (delivers it to the customer), and the data engineer built the kitchen — the plumbing, the gas lines, the walk-in cooler, the delivery dock. Without the kitchen, nobody cooks.

### What Data Engineers Actually Do

| Activity | What It Means | Real Example |
|----------|--------------|---------------|
| **Build data pipelines** | Write code that moves data from point A to point B, transforming it along the way | A PySpark job that reads raw call logs from cloud storage, cleans them, and loads them into a warehouse |
| **Design data models** | Decide how data is structured and organized for querying | Create a star schema so analysts can answer "revenue by campaign" with a simple SQL query |
| **Move data between systems** | Connect different databases, APIs, file systems, and cloud services | Extract customer records from MongoDB, transform with Spark, load into BigQuery |
| **Ensure data quality** | Make sure data is accurate, complete, and trustworthy | Catch duplicate records, validate phone number formats, mask personal information |
| **Orchestrate workflows** | Schedule and manage the order in which pipelines run | Set up Airflow so the ingestion job runs at 6 AM, then the transform, then the report refresh |
| **Optimize performance** | Make slow things fast | Rewrite a Spark job from 4 hours to 20 minutes by fixing a bad join strategy |
| **Monitor and troubleshoot** | Know when things break and fix them quickly | Get a 3 AM alert that data didn't load, find the root cause, fix it before the morning report |
| **Write documentation** | Record how systems work so others can maintain them | Architecture diagrams, data dictionaries, runbooks for incident response |

### Where Data Engineers Work

Data engineers aren't limited to "tech companies." Every industry that generates data needs them.

| Industry | What DEs Build | Example |
|----------|---------------|----------|
| **Finance / Banking** | Transaction pipelines, fraud detection data flows, regulatory reporting | Aggregate millions of daily transactions into risk reports for compliance |
| **Healthcare** | Patient data integration, claims processing, clinical trial data | Combine data from 12 hospital systems into one analytics platform |
| **Retail / E-commerce** | Inventory tracking, recommendation engines, supply chain | Real-time product availability across 500 stores |
| **Telecom** | Call detail records, network performance, customer churn | Process billions of call records daily for billing and usage analysis |
| **Manufacturing** | IoT sensor data, quality control, predictive maintenance | Stream data from 10,000 factory sensors into dashboards |
| **Call Centers** | Call analytics, agent performance, conversion tracking | Combine telephony, payment, and order data into daily performance reports |
| **Government** | Census data, public records, regulatory filings | Build pipelines that process tax filings for audit selection |
| **Media / Advertising** | Ad impression tracking, content recommendations, A/B test data | Process 100M daily ad events to calculate campaign ROI |

> **The takeaway:** If you learn data engineering, you're not locked into one industry. The tools and patterns are the same — only the domain changes.

## 2. What's Expected of a Jr. Data Engineer

**One-line answer:** Here's what employers expect from a junior DE on day one — and what they don't.

Let's be specific. When a company hires a **junior** data engineer, here's what they actually expect.

### Technical Expectations

| Skill | What "Good Enough" Looks Like at Jr. Level |
|-------|--------------------------------------------|
| **SQL** | Write complex queries: JOINs across 3+ tables, window functions, CTEs, aggregations. Read and understand query execution plans. |
| **Python** | Comfortable with data structures, file I/O, working with JSON/CSV, writing functions and classes. Not expected to know ML or advanced libraries. |
| **PySpark** | Read, transform, and write DataFrames. Understand lazy evaluation (Spark doesn't run your code until you ask for a result — it builds a plan first). Know the difference between transformations and actions. |
| **Cloud** | Navigate one cloud platform (GCP, AWS, or Azure). Set up storage, run queries in a cloud warehouse, understand IAM basics. |
| **Git** | Branch, commit, push, pull request, resolve merge conflicts. Follow a team Git workflow. |
| **Linux / CLI** | Navigate the file system, edit files, run scripts, read logs, use environment variables. |
| **Data Modeling** | Understand star schema, fact vs. dimension tables, why normalization matters. Design a simple schema from requirements. |
| **Orchestration** | Understand what a DAG (Directed Acyclic Graph — a sequence of steps where each step runs only after its dependencies complete) is. Read and modify an existing Airflow DAG. Know how scheduling and dependencies work. |
| **CI/CD** (Continuous Integration / Continuous Deployment — automated testing and releasing of code) | Understand the concept. Push code, see a pipeline run, read the output. Not expected to build CI/CD from scratch. |

### Non-Technical Expectations

These matter just as much as coding skills. Many candidates who can code still fail because they lack these.

| Expectation | What It Looks Like |
|-------------|--------------------|
| **Communicate clearly** | "The pipeline failed because the source system changed its date format. I've added format detection. Here's the PR." — not just "I fixed it." |
| **Ask questions early** | "I'm not sure whether to use a full refresh or incremental load here. Can I get a quick review?" — don't spend 3 days on the wrong approach. |
| **Document your work** | Write a README. Add comments to non-obvious code. Keep a runbook. Future-you (and your teammates) will thank you. |
| **Own your mistakes** | "I pushed a bad transform that double-counted revenue. Here's what happened, here's the fix, here's how I'll prevent it next time." |
| **Be reliable** | If you say it'll be done Tuesday, it's done Tuesday. If it won't be, say so Monday — not Wednesday. |
| **Learn continuously** | The tools change every 2-3 years. The engineer who stops learning becomes obsolete. Read, experiment, stay curious. |

> **The reality:** At the Jr. level, companies expect you to be competent with the basics, eager to learn, and easy to work with. Nobody expects you to architect the entire platform on day one.

### What You're NOT Expected to Know (Yet)

Don't panic if you don't know these. They're mid-level and senior DE territory.

| Topic | When You'll Learn It |
|-------|-----------------------|
| Designing a data platform from scratch | After 1-2 years of working on existing platforms |
| Real-time streaming (Kafka, Flink) | Specialization after mastering batch |
| Infrastructure as Code (Terraform, Pulumi) | After you've manually provisioned enough to understand why IaC matters |
| Cost optimization at scale | After you've seen actual cloud bills |
| Managing other engineers | Years down the road (if you want that path) |
| ML model training / fine-tuning | That's a different role (MLE / AI Engineer) |

## 3. What Makes a Good Data Engineer

**One-line answer:** The difference between an average DE and a good one is not technical skill — it's how you think about systems.

There's a difference between a DE who can do the job and a DE who's **good** at it. Here's what separates them.

### The Traits That Matter

| Trait | Average DE | Good DE |
|-------|-----------|----------|
| **Attention to data** | "The pipeline ran successfully" | "The pipeline ran, but row count is 3% lower than yesterday — let me investigate before anyone uses this data" |
| **Defensive coding** | Handles the happy path | Asks: "What if the file is empty? What if the schema changed? What if there are 10x more records than usual?" |
| **Curiosity about the business** | "I loaded the data as requested" | "Why does this report matter? Who uses it? What decisions does it drive?" — understanding the business makes you 10x more effective |
| **Systems thinking** | Fixes the bug | Fixes the bug AND asks: "How did bad data get this far? Where should we add a check so this never reaches Gold?" (Bronze/Silver/Gold are data quality layers — you'll learn these layers in Section 12) |
| **Ownership** | "That's the upstream team's problem" | "I'll file a ticket with the upstream team AND add a validation check on our side so our reports don't break even if they ship bad data" |
| **Simplicity** | Over-engineers a solution because it's "more scalable" | Builds the simplest thing that works. Adds complexity only when the data or requirements demand it. |
| **Communication** | Sends a Slack message: "Pipeline fixed" | Sends: "Pipeline failed at 2 AM due to X. Root cause was Y. I fixed it by Z. Added monitoring to catch this earlier next time. Report data is now current as of 6 AM." |

### The Career Progression

| Level | Focus | Typical Experience |
|-------|-------|--------------------|
| **Jr. Data Engineer** | Build and maintain pipelines designed by others. Fix bugs. Write SQL and PySpark. Learn the systems. | 0-2 years |
| **Data Engineer** | Design and build pipelines end-to-end. Make architectural decisions within a domain. Mentor juniors. | 2-5 years |
| **Sr. Data Engineer** | Own entire data domains. Make technology choices. Design for scale, reliability, and cost. Lead projects. | 5-8 years |
| **Staff / Principal DE** | Set technical direction across teams. Solve cross-cutting problems. Influence org-wide data strategy. | 8+ years |
| **Data Architect** | Design the overall data platform. Evaluate technologies. Set standards and governance. Strategic role. | 8+ years |
| **Engineering Manager** | Lead a team of DEs. Balance technical debt vs. features. Hiring, mentoring, stakeholder management. | 5+ years (management track) |

> **Two career tracks:** You can go deep technically (Staff/Principal/Architect) or move into management. Both are valid. You don't have to decide now.

## 4. The Data Roles: DE, DS, DA, MLE, AI Engineer, BA

**One-line answer:** The data world has six distinct roles — understanding the boundaries prevents you from studying the wrong things.

The data world has a lot of job titles. They overlap, they confuse recruiters, and they mean different things at different companies. Let's clarify.

### The Six Roles at a Glance

| Role | One-Line Definition | Primary Question |
|------|--------------------|-----------------|
| **Data Engineer (DE)** | Builds the systems that move, store, and transform data | "How do we get the right data to the right place, reliably?" |
| **Data Scientist (DS)** | Extracts insights and builds predictive models from data | "What can the data tell us? Can we predict this?" |
| **Data Analyst (DA)** | Analyzes data to answer business questions and create reports | "What happened? Why? What should we do about it?" |
| **ML Engineer (MLE)** | Deploys and operationalizes machine learning models in production | "How do we get this model running reliably at scale?" |
| **AI Engineer** | Builds AI-powered applications using LLMs, agents, and generative AI | "How do we build an application that uses AI to solve this problem?" |
| **Business Analyst (BA)** | Translates business needs into requirements and measures outcomes | "What does the business need? How do we measure success?" |

### Deep Comparison

| Dimension | Data Engineer | Data Scientist | Data Analyst | ML Engineer | AI Engineer | Business Analyst |
|-----------|-------------|---------------|-------------|------------|------------|------------------|
| **Core skill** | Building pipelines & infrastructure | Statistics & modeling | SQL & visualization | ML systems & deployment | LLMs & AI applications | Requirements & process |
| **Main tools** | Spark, Airflow, SQL, Cloud | Python, Jupyter, Scikit-learn, R | SQL, Tableau, Excel, Power BI | PyTorch, MLflow, Docker, K8s | LangChain, OpenAI API, Vector DBs | JIRA, Confluence, Excel, Visio |
| **Primary output** | Pipelines, tables, data APIs | Models, insights, experiments | Reports, dashboards, analysis | Production model services | AI-powered applications | Requirements docs, process maps |
| **Works with data that is** | Raw, messy, large-scale | Clean (thanks to DE), structured | Aggregated, summarized | Feature-engineered, labeled | Unstructured (text, images, docs) | High-level metrics and KPIs |
| **Cares most about** | Reliability, throughput, data quality | Accuracy, statistical significance | Clarity, actionability, trends | Latency, model drift, scaling | User experience, prompt quality | Business value, ROI, adoption |
| **Closest non-tech analogy** | Plumber / Electrician | Research scientist | Accountant / Auditor | Factory engineer | Product designer with AI tools | Translator / Bridge builder |
| **Typical background** | Software engineering, CS | Math, Statistics, Physics | Business, Economics, Analytics | CS, Math, Software engineering | Software engineering, CS | Business, MBA, Domain expertise |

### Clarifying Common Confusion

**"Isn't DE just a subset of DS?"**  
No. They're different disciplines. A data scientist who can't get clean data is stuck. A data engineer who can't build a model is fine — that's not their job. Both are essential.

**"What's the difference between MLE and AI Engineer?"**  
MLE focuses on **traditional ML** — training models (classification, regression, recommendation), managing experiments, deploying to production. AI Engineer focuses on **generative AI** — building applications with LLMs, RAG pipelines, AI agents, prompt engineering. The line is blurring as LLMs become part of ML pipelines.

**"Where does Analytics Engineer fit?"**  
Analytics Engineer is a hybrid of DE + DA. They write SQL transforms (using dbt (a SQL transformation tool for data warehouses), for example) to create clean datasets that analysts can self-serve. Think of them as the person who builds the Gold layer that analysts query.

**"Do I need to know ML to be a DE?"**  
No. But understanding the basics helps. You'll often build pipelines that feed ML models, so knowing what a feature store is or why training data quality matters makes you more effective.

**"Is BA a technical role?"**  
It depends on the company. Some BAs write SQL and build dashboards (overlapping with DA). Others focus purely on requirements, stakeholder management, and process design. Either way, BAs are the bridge between business and technical teams — DEs work with them frequently.

## 5. How These Roles Interact

**One-line answer:** In a real company, the Data Engineer is the central utility — like the electrical grid that every appliance depends on.

In a real company, these roles don't work in isolation. Here's how they connect.

### The Data Flow Across Roles

The diagram below shows a real business scenario. Read it top to bottom: a request starts with the Business Analyst on the left and flows right through each role. Notice that every path passes through the Data Engineer.

> **Notice:** The Data Engineer is at the center. Every other role depends on the DE to make data accessible. This is why DE is one of the most in-demand roles in tech.

### Who Works With Whom — and When

| Interaction | What Happens | How Often |
|-------------|-------------|------------|
| **DE ↔ BA** | BA defines what data is needed, DE builds the pipeline to deliver it. BA validates that the output matches business requirements. | Every new feature or report request |
| **DE ↔ DA** | DA says "I need a table with daily conversion rates by campaign." DE builds the mart. DA queries it for dashboards. | Weekly — DAs are the primary consumers of DE output |
| **DE ↔ DS** | DS needs clean, labeled data for model training. DE builds the feature pipeline. DS provides feedback on data quality and feature definitions. | Per ML project |
| **DE ↔ MLE** | MLE needs a serving pipeline for the trained model. DE builds the data pipeline that feeds the model in production (feature store, batch predictions). | When models go to production |
| **DE ↔ AI Engineer** | AI Engineer needs structured data for RAG (retrieval-augmented generation). DE builds the pipeline that indexes documents and keeps the knowledge base current. | When AI applications need enterprise data |

| **DS ↔ MLE** | DS trains the model and hands it off. MLE packages it, deploys it, monitors it, and handles model drift. | Per model lifecycle |
| **DA ↔ DS** | DA finds patterns ("conversion drops on Fridays"). DS investigates causation and builds predictive models. | Ad-hoc, when analysis reveals opportunities |

### Team Structures You'll Encounter

| Structure | How It Works | Common At |
|-----------|-------------|------------|
| **Centralized data team** | One team has all DEs, DSs, DAs. They serve the whole company. | Mid-size companies |
| **Embedded model** | DEs are embedded in product/business teams. Each team has its own DE. | Large tech companies |
| **Hub and spoke** | Central platform team (DEs) builds shared infrastructure. Domain teams (DA/DS) build on top. | Enterprises |
| **Full-stack data team** | Small team where everyone does a bit of everything. DE writes SQL for reports, DA writes Python for pipelines. | Startups |

> **Your first job** will likely be in one of these structures. Understanding where you fit helps you navigate the organization and be effective faster.

## 6. Day in the Life of a Jr. Data Engineer

**One-line answer:** A typical DE day is 30% debugging, 30% building, 20% meetings, and 20% firefighting — here's what that actually looks like.

Here's a realistic day. Not glamorous, not boring — just real.

**9:00 AM** — Check the Airflow dashboard. Last night's daily pipeline ran successfully. One warning: a source file had 3% more null values than usual. You make a note to investigate.

**9:30 AM** — Standup. Your team lead asks about the new campaign report. You say: "The mart is ready in BigQuery. The DA can start building the dashboard. I still need to add the upsell conversion column — I'll have a PR up by noon."

**10:00 AM** — Write a PySpark transformation to compute upsell acceptance rate grouped by campaign. Add it to the existing Silver-to-Gold transform.

**11:00 AM** — Write a data quality test: upsell_rate must be between 0 and 1. If it's outside that range, something is wrong upstream.

**11:30 AM** — Open a pull request. The senior DE reviews it and suggests using a window function instead of a self-join. You learn something new.

**1:00 PM** — The BA pings you: "The finance team wants to see revenue broken down by VA vs. live agents. Can we add that?" You check the data model — the `channel` field already exists in the mart. You tell the BA: "It's already there. The DA just needs to add a filter in the dashboard."

**2:00 PM** — Deploy the updated pipeline via CI/CD (GitHub Actions). It runs in the staging environment first. Tests pass.

**3:00 PM** — Investigate the null values warning from this morning. Turns out the upstream system changed a field from required to optional. You add a COALESCE (a SQL function that substitutes a default value when a field is missing) with a default value and update the data dictionary.

**4:00 PM** — Document the schema change. Update the runbook with the fix.

**4:30 PM** — Review a junior teammate's Airflow DAG. You notice they hardcoded a file path instead of using a configuration variable. You explain why and show them the pattern.

> **Notice:** Most of the day is writing Python/SQL, debugging data issues, collaborating with other roles (BA, DA, senior DE), and maintaining systems. It's hands-on, problem-solving work.

## 7. ETL vs. ELT — The Two Patterns


> **Key terms:** **ETL** = Extract, Transform, Load (transform data BEFORE loading into the warehouse — clean it first, then store it). **ELT** = Extract, Load, Transform (load raw data first, then transform inside the warehouse — BigQuery/Snowflake are powerful enough to transform after loading). Modern cloud pipelines mostly use ELT because cloud warehouses handle transformation efficiently. Our medallion pipeline (Bronze → Silver → Gold) is an ELT pattern: we load raw data to Bronze (Extract + Load), then transform in Silver/Gold (Transform).


**One-line answer:** ETL transforms data before loading it; ELT loads raw data first and transforms it inside the warehouse.

> Think of ETL like cooking a meal before boxing it for delivery — the box only contains finished food. ELT is like shipping raw ingredients to the restaurant and letting their kitchen prepare it — the raw ingredients are always there if you need to start over.

Every data pipeline follows one of two patterns. You'll use both in your career.

### ETL: Extract → Transform → Load

```
Source  ──→  Transform (outside the warehouse)  ──→  Load into warehouse
```

You pull data out, clean/transform it using an external tool (Spark, Python, Glue), then load the **clean** result into the warehouse. The warehouse only ever sees finished data.

### ELT: Extract → Load → Transform

```
Source  ──→  Load raw into warehouse  ──→  Transform inside the warehouse
```

You dump the raw data into the warehouse first, then use the warehouse's own SQL engine to transform it. The raw data is always preserved.

### Side-by-Side Comparison

| Aspect | ETL | ELT |
|--------|-----|-----|
| **Transform happens** | Before loading (external compute) | After loading (warehouse compute) |
| **Raw data in warehouse?** | No — only clean data arrives | Yes — raw data lands first |
| **Compute** | External: Spark cluster, Python server, AWS Glue | Warehouse engine: BigQuery, Redshift, Snowflake |
| **Audit trail** | Harder — raw data may not be preserved | Easy — raw data is always in the warehouse |
| **Cost model** | Pay for external compute cluster | Pay for warehouse processing |
| **Best for** | Complex transforms, ML pipelines, multi-target loads | SQL-heavy transforms, analyst self-service, cloud warehouses |
| **Legacy tools** | Informatica, SSIS, Talend, DataStage | — |
| **Modern tools** | PySpark, AWS Glue, Google Dataflow | dbt, BigQuery SQL, Snowflake SQL |

### In Practice: Most Pipelines Blend Both

Modern pipelines are actually **E-LT** — a hybrid:

1. **Extract** from source systems
2. **Light transform** on the way in (schema enforcement, format conversion, dedup)
3. **Load** raw data into a Bronze layer
4. **Heavy transform** inside the warehouse using SQL or Spark

> **In this program:** You'll use PySpark for complex transforms (ETL-style) and BigQuery for analytical queries (ELT-style). We cover this hands-on in M01.

## 8. Cloud Platforms: AWS vs. GCP vs. Azure

**One-line answer:** The cloud is renting a massive computer in someone else's data center instead of buying your own.

Every DE needs to know at least one cloud. In this program we use **GCP**, but you'll encounter AWS and Azure in interviews and on the job. The concepts are the same — only the service names differ.

### DE-Relevant Services — Head to Head

| Category | AWS | GCP | Azure | What a DE Uses It For |
|----------|-----|-----|-------|----------------------|
| **Object Storage** | S3 | Cloud Storage (GCS) | Blob Storage / ADLS Gen2 | Store raw data files (CSV, JSON, Parquet) — your Bronze layer lives here |
| **Data Warehouse** | Redshift | **BigQuery** | Synapse Analytics | Run SQL on massive datasets — your Gold layer and reporting |
| **Managed Spark** | EMR | **Dataproc** | HDInsight / Synapse Spark | Run PySpark jobs at scale — your processing engine |
| **Serverless ETL** | Glue | Dataflow | Data Factory | Managed ETL without provisioning servers — auto-scaling transforms |
| **Orchestration** | MWAA (managed Airflow) | **Cloud Composer** (managed Airflow) | Data Factory Pipelines | Schedule and coordinate pipeline steps — your DAGs run here |
| **Streaming** | Kinesis | Pub/Sub + Dataflow | Event Hubs + Stream Analytics | Process real-time data (click streams, IoT, logs) as it arrives |
| **NoSQL** | DynamoDB | Firestore / Bigtable | Cosmos DB | Store semi-structured data, key-value lookups, time-series at scale |
| **SQL Database** | RDS / Aurora | Cloud SQL / AlloyDB | Azure SQL / SQL MI | Managed relational databases for applications (OLTP source systems) |
| **Data Catalog** | Glue Data Catalog | Data Catalog | Purview | Track what data exists, where it lives, who owns it — data governance |
| **IAM** | IAM + Roles | IAM + Service Accounts | Entra ID + RBAC | Control who can access what — critical for PII/PCI compliance |
| **CLI** | `aws` | `gcloud` / `gsutil` / `bq` | `az` | Manage cloud resources from terminal — scripting and automation |
| **Serverless Compute** | Lambda | Cloud Functions | Azure Functions | Run small pieces of code on demand — trigger transforms on file upload |
| **Containers** | EKS | GKE | AKS | Run containerized pipelines — package your Spark/Airflow in Docker |
| **Monitoring** | CloudWatch | Cloud Monitoring | Azure Monitor | Track pipeline health, set alerts for failures, measure performance |
| **CI/CD** | CodePipeline | Cloud Build | Azure DevOps | Automated testing and deployment of pipeline code |
| **Cost Management** | Cost Explorer | Billing + Recommender | Cost Management | Track cloud spend — BigQuery charges per query, Dataproc per cluster-hour |

> **Bold** = what you'll use in this program

### Advantages and Disadvantages

| Cloud | Advantages | Disadvantages |

## 9. Core Concepts: OLTP vs OLAP, ACID vs BASE, CAP Theorem

**One-line answer:** These four concepts — OLTP, OLAP, ACID, and CAP — show up in every DE interview and govern every architecture decision.

Three foundational concepts that come up in every DE interview and every architecture decision. Know these cold.

### OLTP vs. OLAP

Two fundamentally different ways to use a database. As a DE, you move data FROM OLTP systems INTO OLAP systems.

> OLTP is the cash register — it records one transaction at a time, fast and accurate. OLAP is the end-of-month financial report — it reads everything, takes time, but produces insight.

| | OLTP (Online Transaction Processing) | OLAP (Online Analytical Processing) |
|---|---|---|
| **Purpose** | Run the business — process transactions | Analyze the business — answer questions |
| **Operations** | INSERT, UPDATE, DELETE (many small writes) | SELECT with aggregations (few large reads) |
| **Users** | Application users, customers, cashiers | Analysts, data scientists, executives |
| **Query pattern** | "Get order #12345" (one row, fast) | "Total revenue by region for Q1" (millions of rows, aggregated) |
| **Schema** | Normalized (3NF) — minimize redundancy | Denormalized (star schema) — optimize for reads |
| **Data freshness** | Real-time — current state of the business | Near-real-time to daily — historical + current |
| **Examples** | PostgreSQL, MySQL, Oracle, SQL Server | BigQuery, Redshift, Snowflake, ClickHouse |
| **Row count** | Thousands to millions | Millions to billions |
| **Typical query time** | Milliseconds | Seconds to minutes |
| **Who manages it** | Application developers / DBAs | **Data Engineers** |

```
OLTP (source)  ──→  DE Pipeline (ETL/ELT)  ──→  OLAP (warehouse)  ──→  Reports
PostgreSQL          PySpark + Airflow            BigQuery               Dashboards
"Process the        "Move and transform          "Analyze the           "Show me the
 order"              the data"                    trends"                numbers"
```

> **Your role as a DE:** You don't build the OLTP system (that's the app team). You don't query the OLAP system for insights (that's the analyst). You build the **pipeline between them** — extract from OLTP, transform, load into OLAP.

### ACID vs. BASE

Two different approaches to data consistency. Understanding when to use each is a core DE skill.

**ACID** — the traditional database guarantee (PostgreSQL, MySQL, Delta Lake):

> Think of ACID like a bank vault door — it either opens completely or stays shut. No halfway. Your $100 transfer either moves fully or not at all.

| Property | What It Means | Example |
|----------|--------------|---------|
| **Atomicity** | A transaction either fully completes or fully rolls back. No partial writes. | Transfer $100: debit AND credit both happen, or neither does |
| **Consistency** | The database always moves from one valid state to another. Rules are never violated. | Account balance can never go negative (if that's a constraint) |
| **Isolation** | Concurrent transactions don't interfere with each other. | Two people buying the last item — only one succeeds |
| **Durability** | Once committed, data survives crashes, power failures, etc. | After "commit," the data is on disk — permanently |

**BASE** — the distributed systems trade-off (MongoDB, Cassandra, DynamoDB):

| Property | What It Means | Example |
|----------|--------------|---------|
| **Basically Available** | The system always responds, even if the data might be stale | Your social media feed loads, but might not show the latest post yet |
| **Soft state** | Data may change over time even without new input (as replicas sync) | Two replicas might briefly show different values |
| **Eventually consistent** | All replicas will converge to the same value — eventually, not instantly | After you post a photo, some friends see it immediately, others see it 2 seconds later |

### When to Use Each

| Scenario | Use | Why |
|----------|-----|-----|
| Bank transactions | **ACID** | You cannot have $100 disappear between accounts |
| E-commerce orders | **ACID** | Inventory must be accurate — can't sell what you don't have |
| Social media likes | **BASE** | It's OK if the count is off by 1 for a few seconds |
| IoT sensor data | **BASE** | Volume is massive, occasional inconsistency is acceptable |
| Data warehouse loads | **ACID** (Delta Lake) | Reports must be accurate — partial loads cause wrong numbers |
| Call center call logs | **ACID for pipeline** | You need exactly-once processing — no duplicate revenue |

> **DE relevance:** This is why **Delta Lake** exists. Regular data lakes (just files on S3/GCS) have no ACID guarantees — a failed write can leave corrupted data. Delta Lake adds ACID transactions to your data lake. You'll learn this in M09.

### CAP Theorem

> Think of CAP as choosing two items from a three-item menu — you can have any combination of two, but never all three simultaneously.

In a distributed system, you can only guarantee **two out of three**:

```
        Consistency
           /\
          /  \
         /    \
        / Pick \
       /  Two   \
      /    !!     \
     /______________\
Availability    Partition Tolerance
```

| Property | What It Means |
|----------|--------------|
| **Consistency** | Every read gets the most recent write (all nodes have the same data) |
| **Availability** | Every request gets a response (system never says "I'm down") |
| **Partition Tolerance** | System keeps working even when network splits nodes apart |

**You always need Partition Tolerance** in distributed systems (networks fail — that's reality). So the real choice is:

| Choice | What You Get | What You Sacrifice | Example Systems |
|--------|-------------|-------------------|----------------|
| **CP** (Consistency + Partition Tolerance) | Always-correct data | May reject requests during network issues | HBase, MongoDB (default), Redis |
| **AP** (Availability + Partition Tolerance) | Always responds | May return stale data briefly | Cassandra, DynamoDB, CouchDB |
| **CA** (Consistency + Availability) | Perfect but fragile | Breaks on network partition — not practical for distributed | Single-node PostgreSQL, traditional RDBMS |

### CAP in the DE World

| System | CAP Choice | Why It Matters to You |
|--------|-----------|----------------------|
| **BigQuery** | CP | Your queries always return correct results (no stale data) |
| **Cloud Storage (GCS/S3)** | AP | Uploads are eventually consistent — a file you just wrote might not show in a list for a moment |
| **Delta Lake** | CP | ACID transactions ensure your pipeline writes are correct |
| **Kafka** | AP | Messages are always accepted, consumers may lag briefly |
| **MongoDB** | Configurable (CP or AP) | Depends on read/write concern settings |

> **Interview tip:** When asked about CAP theorem, don't just recite the definition. Say: "In our data pipeline, we use Delta Lake which is CP — we need consistency because reports must be accurate. But our ingestion layer uses GCS which is AP — we can tolerate brief inconsistency on file listings because Airflow retries handle it."

## 10. Data Classification: PII, PCI, PHI, and Sensitivity Levels

**One-line answer:** If you mishandle personal data, your company faces fines, lawsuits, and lost trust — knowing the categories prevents this.

As a DE, you're the last line of defense before sensitive data reaches reports, dashboards, or downstream systems. If you don't handle it correctly, your company faces fines, lawsuits, and lost trust.

### Data Sensitivity Categories

| Category | Full Name | What It Covers | Regulation | Violation Consequence |
|----------|-----------|---------------|------------|----------------------|
| **PII** | Personally Identifiable Information | Name, email, phone, address, SSN, date of birth — anything that identifies a specific person | GDPR, CCPA, state privacy laws | Fines up to 4% of global revenue (GDPR), class action lawsuits |
| **PCI** | Payment Card Industry Data | Credit/debit card numbers, CVV, expiration dates, cardholder name | PCI DSS | Fines $5K–$100K/month, lose ability to process cards |
| **PHI** | Protected Health Information | Medical records, diagnoses, prescriptions, insurance IDs, lab results | HIPAA | Fines $100–$50K per violation, criminal penalties |
| **Financial** | Financial / SOX Data | Revenue figures, audit trails, internal financial reports | SOX, SEC regulations | Criminal penalties for executives, company delisting |
| **Internal** | Internal / Confidential | Employee data, salaries, strategic plans, trade secrets | Company policy, NDAs | Termination, lawsuits |
| **Public** | Public Data | Published reports, marketing materials, open datasets | None | None — it's already public |

### What This Looks Like in Your Pipeline

| Raw Data (Bronze) | Classification | What You Must Do (Silver) |
|-------------------|---------------|--------------------------|
| `"phone": "313-555-9876"` | **PII** | Mask: `"phone": "313-***-****"` or hash it |
| `"email": "jane@email.com"` | **PII** | Remove from reporting tables, keep only in restricted access |
| `"card_number": "4111111111111111"` | **PCI** | **Never store.** If it arrives in your pipeline, quarantine and alert immediately |
| `"ssn": "123-45-6789"` | **PII (high)** | **Never store in analytics.** Tokenize or drop entirely |
| `"diagnosis": "Type 2 Diabetes"` | **PHI** | Encrypt, restrict access, audit every query |
| `"campaign_name": "BullsEye Pro"` | **Internal** | OK in internal reports, not in external-facing data |
| `"total_calls": 1452` | **Aggregated** | Safe — no individual is identifiable from aggregated stats |

> **Key principle:** Aggregate data is almost always safe. Individual-level data is where the risk lives. When in doubt, aggregate or mask.

### The DE's Responsibility

| Stage | What You Do |
|-------|------------|
| **Bronze (raw)** | Data arrives with PII/PCI. Store it in a restricted bucket/table with limited access. Log who accesses it. |
| **Silver (clean)** | Mask, hash, or remove PII. Validate that no PCI data leaked through. Apply column-level access controls. |
| **Gold (reporting)** | PII should be gone. Only aggregated or anonymized data reaches dashboards and analysts. |
| **Access control** | Set up IAM roles so only authorized users can query Silver tables with PII. Gold tables are open to analysts. |
| **Audit trail** | Log every access to sensitive data. If a regulator asks "who saw this data?", you need an answer. |

### Common Mistakes (That Get People Fired)

| Mistake | Why It's Bad | How to Prevent |
|---------|-------------|----------------|
| Logging PII to application logs | Logs are often unprotected and widely accessible | Scrub PII from log messages, use IDs instead of names |
| Storing credit card numbers "temporarily" | PCI DSS doesn't care if it's temporary — if you stored it, you're in scope | Never accept raw card data in your pipeline |
| Copying production data to dev/staging | Dev environments usually have weaker security | Use synthetic or masked data for development |
| Emailing a CSV with customer names | Email is not encrypted, not access-controlled | Use secure file shares with access logging |
| No column-level access control | Everyone on the team can see SSNs and salaries | Use BigQuery column policies, or separate sensitive columns into restricted tables |
| Joining datasets that re-identify people | Combining zip + age + gender can uniquely identify someone | Test for re-identification risk before publishing joined datasets |

### Data Classification in Our Call Center Project

| Field | Classification | Handling |
|-------|---------------|----------|
| `caller_phone` | PII | Masked in Silver/Gold: `313-***-****` |
| `customer.email` | PII | Removed from Gold layer entirely |
| `customer.first_name`, `last_name` | PII | Available in Silver (restricted), not in Gold |
| `customer.billing_address` | PII | Hashed or removed in Silver |
| `card_type` (Visa, MC) | Not PCI (it's the type, not the number) | OK in reporting |
| `auth_code` | PCI-adjacent | Keep in Silver (restricted), not in Gold |
| `call_id`, `dnis`, `disposition` | Internal | Safe for all reporting |
| `total_revenue` (aggregated) | Internal | Safe — no individual is identifiable |

## 11. Why Pipelines? (Not One Big Program)

**One-line answer:** A pipeline is a script that can recover from failure, track what it did, and run again safely — a plain script cannot.

A common question from new DEs: "Why can't I just write one Python script that reads the data, cleans it, and loads it into the database? Why do I need all these tools?"

### The One-Script Approach

```python
# The "just write one script" approach
data = read_csv("calls.csv")          # Step 1: Read
clean = remove_nulls(data)             # Step 2: Clean
transformed = add_columns(clean)       # Step 3: Transform
load_to_database(transformed)          # Step 4: Load
print("Done!")
```

**This works... until it doesn't.**

### Why It Breaks in Production

| Problem | What Happens | Pipeline Solution |
|---------|-------------|-------------------|
| **Step 3 fails** | The whole script crashes. Steps 1-2 are wasted. You restart from scratch. | Each step is independent. Retry only the failed step. |
| **Data grows from 1K to 100M rows** | Script runs out of memory. Your laptop can't handle it. | Pipeline runs on a Spark cluster — scales horizontally. |
| **You need to run it every day at 6 AM** | You set a cron job. It fails silently on weekends. Nobody notices until Monday. | Airflow schedules it, retries on failure, and alerts you. |
| **Another team needs the same clean data** | You copy-paste the cleaning code. Now there are two versions. They drift apart. | Silver layer is the single source of truth. Everyone reads from it. |
| **The source format changes** | The whole script breaks. You can't tell which step caused it. | Bronze layer catches the raw data as-is. Schema enforcement in Silver catches the change. |
| **Boss asks "what did yesterday's report show?"** | You have no idea — the data was overwritten. | Delta Lake time travel: query any historical version. |
| **Auditor asks "who touched this data?"** | No logs. No trail. | Airflow logs every run. Delta Lake logs every write. |

### The Pipeline Approach

```
Source → Ingestion → Bronze → Silver → Gold → Reports
         (Airflow)   (raw)   (clean)  (marts)  (BigQuery)
```

Each step is:
- **Independent** — fails and retries without affecting other steps
- **Observable** — you can see what ran, when, and whether it succeeded
- **Reusable** — multiple consumers can read from the same layer
- **Auditable** — every transformation is logged and versible
- **Scalable** — runs on a cluster, not your laptop

### The Mental Model

| Single Script | Pipeline |
|--------------|----------|
| Cooking a meal start to finish, alone, in one take | A restaurant kitchen with stations: prep, grill, plating, service |
| If anything goes wrong, start over | If the grill burns a steak, re-fire just that steak |
| Works for dinner at home | Required for serving 500 customers a night |

> **The rule:** If data is small, one-time, and only you use it — a script is fine. If data is large, recurring, and others depend on it — you need a pipeline. In production, it's always a pipeline.

### Pipeline vs. Single Batch Job — Why Pipelines Win

| | Single Batch Job | Data Pipeline |
|---|---|---|
| **Architecture** | One script does everything: read → transform → write | Separate stages: ingest → store → process → serve, each independent |
| **Failure handling** | Entire job fails — restart from scratch, lose hours of work | One stage fails — others continue, retry only the failed stage, raw data is safe |
| **Scalability** | Limited by one machine's memory and CPU | Each stage scales independently (add more Spark nodes, bigger warehouse) |
| **Monitoring** | "Did it finish?" — check the output file | Per-stage metrics: ingestion lag, transform duration, data quality scores, serve latency |
| **Data recovery** | If something goes wrong, re-run the whole job (if you still have the source data) | Bronze layer preserves raw data forever — reprocess from any point in time |
| **Schema changes** | Upstream table adds a column → entire job breaks | Schema change detected at ingestion → alert fires → only transform logic needs updating |
| **Multiple consumers** | One output format — everyone gets the same flat file | Gold layer serves different marts for different teams (finance, marketing, ops) |
| **Debugging** | "Something is wrong with the numbers" — search through 500 lines of SQL | Check each layer: Is Bronze correct? Is Silver transform right? Is Gold aggregation accurate? |
| **Team collaboration** | One person owns the script — bus factor of 1 | Different engineers own different stages — parallel development, clear ownership |
| **Reprocessing** | Re-run everything from scratch | Replay from Bronze (raw data preserved), reprocess only Silver→Gold |
| **Testing** | Test the whole monolith or nothing | Test each stage independently — unit test transforms, integration test stage boundaries |
| **Cost** | Pay for a big server that's idle 23 hours/day | Pay per use — serverless ingestion, auto-scaling clusters, query-based warehouse billing |

> **The real-world pattern:** Every pipeline example above follows the same Bronze → Silver → Gold structure. The tools change (Kinesis vs. API, Glue vs. EMR, Redshift vs. DynamoDB), but the pattern is universal. Learn the pattern once, apply it everywhere.

> **Interview phrasing:** "In my training, I built a multi-stage data pipeline using the medallion architecture — Bronze for raw ingestion, Silver for cleaning and conforming, Gold for business-ready marts. Each stage was independently testable, retryable, and scalable. Compared to a single batch job, this gave us data recovery through Bronze, isolated failure handling, and the ability to serve different consumers from the same pipeline."

## 13. Data Lake vs. Data Warehouse vs. Data Lakehouse

**One-line answer:** Data lakes store everything cheaply, warehouses make it queryable, and lakehouses combine both — that's the 20-year evolution.

Three terms you'll hear constantly. They represent the evolution of how companies store and analyze data.

### The Three Architectures

```
DATA WAREHOUSE (2000s)              DATA LAKE (2010s)                DATA LAKEHOUSE (2020s)
┌─────────────────────┐        ┌─────────────────────┐        ┌─────────────────────┐
│  Structured data     │        │  Everything: files,  │        │  Everything + ACID   │
│  Tables, SQL         │        │  JSON, CSV, images   │        │  Tables on top of    │
│  Schema-on-write     │        │  Schema-on-read      │        │  files in the lake   │
│  Fast queries        │        │  Cheap storage       │        │  Best of both        │
│                     │        │                     │        │                     │
│  Redshift, BigQuery  │        │  S3/GCS + Spark      │        │  Delta Lake, Iceberg │
│  Snowflake           │        │  HDFS                │        │  on S3/GCS           │
└─────────────────────┘        └─────────────────────┘        └─────────────────────┘
     Expensive, rigid              Cheap, messy                 Cheap, structured
```

### Head-to-Head Comparison

> Schema-on-write = decide what the form looks like before you fill it in. Schema-on-read = dump all the papers in a box, figure out what they mean when you need them.

| | Data Warehouse | Data Lake | Data Lakehouse |
|---|---|---|---|
| **What it stores** | Structured data only (tables, rows, columns) | Everything: structured, semi-structured (JSON), unstructured (images, logs) | Everything — but with table structure layered on top |
| **Schema** | Schema-on-write: define the table structure before loading | Schema-on-read: dump anything in, figure out structure when you query | Schema-on-write with schema evolution (can change over time) |
| **Storage cost** | Expensive (proprietary format, pay for storage + compute) | Cheap (files on S3/GCS — pennies per GB) | Cheap (same files on S3/GCS, but with metadata layer) |
| **Query speed** | Fast (optimized engine) | Slow on raw files (Spark job reads everything) | Fast (table format optimizes reads with metadata, partitioning, Z-ordering) |
| **ACID transactions** | Yes | No — a failed write can corrupt your data | Yes (Delta Lake, Iceberg add ACID to the lake) |
| **Time travel** | Limited or none | None | Yes — query any historical version of the data |
| **Who uses it** | Analysts, BI tools | Data scientists, ML engineers | Everyone — analysts AND data scientists |
| **Examples** | BigQuery, Redshift, Snowflake | S3 + Spark, HDFS + Hive | Delta Lake on GCS, Apache Iceberg on S3 |
| **Era** | 2000s–present | 2010s–present | 2020s–present |

### Advantages and Disadvantages

| Architecture | Advantages | Disadvantages |
|---|---|---|
| **Data Warehouse** | Fast SQL queries, strong data governance, ACID transactions, mature tooling (20+ years), BI tools plug in natively | Expensive at scale, structured data only (no images, logs, JSON), rigid schemas (hard to change), vendor lock-in (migrating off Snowflake is painful) |
| **Data Lake** | Cheap storage (pennies/GB), stores any format (CSV, JSON, images, video), flexible schema, great for ML/data science, no vendor lock-in (just files) | No ACID (failed writes corrupt data), becomes "data swamp" without governance, slow queries on raw files, no schema enforcement, hard to guarantee data quality |
| **Data Lakehouse** | Cheap storage + ACID transactions, schema evolution, time travel, serves both analysts and data scientists, open formats (not locked to one vendor) | Newer technology (less mature), requires Delta Lake/Iceberg expertise, more complex architecture than a simple warehouse, tooling still evolving |

### Real-World Use Cases

| Use Case | Best Architecture | Why |
|---|---|---|
| Bank regulatory reporting (daily compliance) | **Data Warehouse** | Strict governance, fast SQL, ACID required, all data is structured tables |
| Startup analytics dashboard (100K rows) | **Data Warehouse** | Small scale, SQL-first, don't need a lake, fast time-to-insight |
| ML training on raw images + text + tabular data | **Data Lake** | Need to store unstructured data (images, text files), data scientists access directly with Spark/Python |
| IoT sensor data (billions of events, store everything) | **Data Lake** | Massive volume, cheap storage critical, varied formats (JSON, binary), ML workloads |
| Media company storing video + metadata + user behavior | **Data Lake** | Video files can't go in a warehouse, need flexible storage for mixed formats |
| Enterprise analytics platform (500+ users, mixed data) | **Data Lakehouse** | Cheap storage for volume, ACID for reliability, SQL for analysts, flexible access for data scientists — serves everyone |
| Your capstone project | **Data Lakehouse** | Raw JSON/CSV in GCS (lake) + Delta Lake for ACID (lakehouse) + BigQuery for SQL (warehouse serving layer) |

### The Evolution Story

**Phase 1 — Data Warehouse:** Companies put structured data into expensive warehouses. Analysts loved it (fast SQL queries). But: you couldn't store unstructured data (logs, JSON, images), it was expensive at scale, and loading data required rigid schemas.

**Phase 2 — Data Lake:** "Just dump everything into S3/HDFS!" Storage was cheap, and you could store anything. But: data lakes became "data swamps" — no organization, no quality checks, no ACID transactions. A failed Spark job could leave corrupted files that broke downstream reports.

**Phase 3 — Data Lakehouse:** "What if we kept the cheap storage of a lake but added the reliability of a warehouse?" Delta Lake and Apache Iceberg add ACID transactions, schema enforcement, and time travel on top of plain files in cloud storage. You get the best of both worlds.

### Where We Are in This Program

| Layer | Technology | Architecture |
|-------|-----------|-------------|
| **Storage** | GCS (Cloud Storage) | Data Lake — cheap, scalable file storage |
| **Table Format** | Delta Lake + Iceberg | Lakehouse — ACID, schema evolution, time travel |
| **Analytics** | BigQuery | Warehouse — fast SQL for analysts |
| **Processing** | PySpark on Dataproc | Lake processing — distributed transforms |

> **You're learning the lakehouse pattern.** Raw files land in GCS (lake). Delta Lake / Iceberg add structure and reliability (lakehouse). BigQuery serves fast analytical queries (warehouse). This is the modern architecture most companies are adopting — and it's what you'll build in the capstone.

> **Interview phrasing:** "I've worked with the lakehouse architecture — raw data in cloud storage with Delta Lake for ACID transactions and schema evolution, and BigQuery for the serving layer."

## 14. Data Formats: CSV, JSON, Parquet, Avro

**One-line answer:** The file format you choose affects storage cost, query speed, and compatibility — Parquet wins for analytics.

As a DE, you don't just move "data" — you move **files in specific formats**. The format you choose affects storage cost, query speed, and compatibility. You'll work with all four of these.

### The Four Formats at a Glance

| Format | Type | Human Readable? | Best For | You'll Use It In |
|--------|------|----------------|----------|-----------------|
| **CSV** | Row-based, text | Yes — open in Excel or any editor | Simple tabular data, imports/exports, small datasets | M03 (SQL), M06 (PySpark ingest) |
| **JSON** | Row-based, text | Yes — structured key-value pairs | Nested/semi-structured data, API responses, MongoDB exports | M06 (call records), M16 (capstone) |
| **Parquet** | Columnar, binary | No — need special tools to read | Analytics, warehouses, large datasets, column-heavy queries | M06 (PySpark output), M08 (BigQuery), M09 (Delta Lake) |
| **Avro** | Row-based, binary | No — binary with embedded schema | Streaming, schema evolution, Kafka messages | M05 (Hadoop context) |

### Advantages and Disadvantages

| Format | Advantages | Disadvantages |
|---|---|---|
| **CSV** | Universal — every tool reads it (Excel, Python, Spark, databases), human-readable and easy to debug, simple to create and edit, no special libraries needed | No schema (column types are guessed), no nested data (flat tables only), large files (no compression), delimiter conflicts (commas in data break parsing), no metadata (column names can be wrong or missing) |
| **JSON** | Supports nested/hierarchical data naturally, human-readable and self-describing, standard for APIs and web services, flexible schema (each record can differ) | Large files (keys repeated in every row — "customer_name" stored 100K times), slow for analytics (row-based, can't skip columns), no built-in schema enforcement (anything goes), parsing slower than binary formats |
| **Parquet** | Excellent compression (5-10x smaller than CSV), columnar (skip unused columns = fast analytics), embedded schema with data types, splittable for parallel processing (Spark reads partitions independently) | Not human-readable (binary — can't open in a text editor), slower for single-row lookups (must reconstruct from columns), requires Spark/PyArrow/BigQuery to read, write overhead (slower to create than CSV) |
| **Avro** | Compact binary format, strong schema evolution (add/remove fields safely), schema travels with the data file, fast serialization/deserialization, splittable | Not human-readable, less common than Parquet for analytics workloads, row-based (not optimal for column-heavy analytical queries), smaller ecosystem than Parquet |

### Real-World Use Cases

| Use Case | Best Format | Why |
|---|---|---|
| Export data for a business user to open in Excel | **CSV** | Universal — every spreadsheet tool reads it instantly |
| Share a small dataset with a colleague | **CSV** | Simple, no special tools, anyone can open it |
| API response from a REST endpoint | **JSON** | Standard web format, supports nested objects (address inside customer) |
| MongoDB export of call records | **JSON** | MongoDB stores documents as JSON natively |
| Store 100M call records for analytical queries | **Parquet** | 10x smaller than CSV, columnar = fast aggregations (SUM, AVG, GROUP BY) |
| Data lake Bronze layer (preserve raw data) | **Parquet** | Efficient storage, schema enforcement, splittable for distributed processing |
| Streaming events through Kafka | **Avro** | Schema evolution handles changing event formats, compact, fast serialization |
| Data exchange between Spark jobs in a pipeline | **Parquet** | Native Spark format, efficient columnar reads, preserves schema |
| Config files or small lookup tables (<100 rows) | **CSV or JSON** | Simple, human-readable, easy to edit by hand |
| Archive historical data for compliance | **Parquet** | Massive compression saves storage cost, schema preserved for future reads |

### Row-Based vs. Columnar — Why It Matters

```
ROW-BASED (CSV, JSON, Avro)              COLUMNAR (Parquet)

Row 1: id=1, name="Alice", age=30        Column "id":   [1, 2, 3, 4, ...]
Row 2: id=2, name="Bob",   age=25        Column "name": ["Alice", "Bob", ...]
Row 3: id=3, name="Carol", age=35        Column "age":  [30, 25, 35, ...]
Row 4: id=4, name="Dave",  age=28        
```

| Query | Row-Based | Columnar |
|-------|----------|---------|
| `SELECT * FROM users WHERE id = 3` | Fast — read one row | Slower — must reconstruct row from columns |
| `SELECT AVG(age) FROM users` | Slow — must read ALL columns of ALL rows | Fast — reads only the `age` column, skips everything else |
| `SELECT name, age FROM users WHERE age > 30` | Reads entire rows, discards unused columns | Reads only `name` and `age` columns — much less I/O |

> **Rule of thumb:** Row-based formats are good for writing and transactional access (get one record). Columnar formats are good for analytical queries (aggregate millions of records). DEs write Parquet because analysts run analytical queries.

### Detailed Comparison

| | CSV | JSON | Parquet | Avro |
|---|---|---|---|---|
| **Storage size** (1M rows) | ~100 MB | ~150 MB (keys repeated per row) | ~20 MB (compressed, columnar) | ~30 MB (binary, compact) |
| **Schema** | None — just headers | Flexible — nested objects, arrays | Embedded — typed columns, metadata | Embedded — schema stored with data |
| **Compression** | Poor (text, redundant) | Poor (keys repeat every row) | Excellent (same-type values compress well) | Good (binary, schema-aware) |
| **Nested data** | No (flat tables only) | Yes (objects inside objects) | Yes (Parquet supports nested structs) | Yes |
| **Splittable** (can Spark read in parallel?) | Yes | Yes (if newline-delimited) | Yes | Yes |
| **Schema evolution** | No — add a column and old files break | No formal mechanism | Yes — add columns, readers ignore unknown | Yes — designed for schema evolution |
| **Common tools** | Excel, pandas, any text editor | APIs, MongoDB, web apps | Spark, BigQuery, Athena, Hive | Kafka, Hadoop, Spark |

### Compression Formats You'll See

| Compression | Speed | Ratio | Splittable? | Common With |
|-------------|-------|-------|-------------|-------------|
| **Snappy** | Very fast compress/decompress | Moderate (~2-4x) | Yes | Parquet default, PySpark |
| **GZIP** | Slower | High (~5-8x) | No | CSV exports, web data |
| **BZIP2** | Slowest | Highest (~6-10x) | Yes | Archival, maximum compression |
| **LZ4** | Fastest | Lower (~2-3x) | Yes | Real-time, low-latency |
| **ZSTD** | Fast | High (~5-8x) | Yes | Modern default, Delta Lake |

> **In this program:** You'll read CSV and JSON (source data), transform with PySpark, and write Parquet (output). This is the standard pattern: **ingest as text, process in memory, persist as Parquet.**

### The Call Center Dataset Formats

| File | Format | Why This Format |
|------|--------|----------------|
| `calls.json` | JSON | Nested customer objects, flexible schema from VA platform |
| `orders.csv` | CSV | Flat tabular data from order management system |
| `payments.xml` | XML | Payment gateway legacy format |
| `products.csv` | CSV | Simple lookup table, rarely changes |
| `campaigns.csv` | CSV | Simple lookup table |

## 16. Data Modeling: Star Schema, Facts, Dimensions, Normalization

**One-line answer:** Data modeling is how you organize your data so queries are fast and intuitive instead of painful.

Data modeling is how you decide to **organize** your data. The right model makes queries fast and intuitive. The wrong model makes everything painful. This is a preview — you'll go deep in M04.

### What is a Data Model?

A data model is the blueprint for how data is structured in a database: what tables exist, what columns they have, and how they relate to each other. It's the difference between dumping everything into one giant spreadsheet vs. organizing it into clear, connected tables.

### Normalization — Eliminating Redundancy

Normalization is the process of organizing tables to **reduce duplication**. There are formal "normal forms" (1NF, 2NF, 3NF), but the intuition is simple: **don't store the same fact in two places.**

**The Problem — One Big Table (Unnormalized)**

| call_id | customer_name | customer_phone | campaign_name | campaign_client | product_name | product_price | amount |
|---------|--------------|----------------|---------------|-----------------|-------------|---------------|--------|
| C001 | Alice Smith | 313-555-1234 | BullsEye Pro | Acme Corp | Widget A | 29.99 | 29.99 |
| C002 | Alice Smith | 313-555-1234 | BullsEye Pro | Acme Corp | Widget B | 49.99 | 49.99 |
| C003 | Bob Jones | 248-555-5678 | SpringSale | Acme Corp | Widget A | 29.99 | 29.99 |

**What's wrong here?**
- "Alice Smith" and her phone number are stored in every row she appears in. If she changes her phone number, you have to update every row.
- "BullsEye Pro" and "Acme Corp" are repeated. If the campaign name changes, you update dozens of rows.
- "Widget A" and its price appear multiple times. If the price changes, which rows do you update?

**The Fix — Normalize Into Separate Tables**

**customers table:**

| customer_id | name | phone |
|-------------|------|-------|
| 1 | Alice Smith | 313-555-1234 |
| 2 | Bob Jones | 248-555-5678 |

**campaigns table:**

| campaign_id | campaign_name | client |
|-------------|--------------|--------|
| 10 | BullsEye Pro | Acme Corp |
| 11 | SpringSale | Acme Corp |

**products table:**

| product_id | product_name | price |
|------------|-------------|-------|
| 100 | Widget A | 29.99 |
| 101 | Widget B | 49.99 |

**calls table:**

| call_id | customer_id | campaign_id | product_id | amount |
|---------|-------------|-------------|------------|--------|
| C001 | 1 | 10 | 100 | 29.99 |
| C002 | 1 | 10 | 101 | 49.99 |
| C003 | 2 | 11 | 100 | 29.99 |

**Now:** Alice's phone number is stored once. Change it in one place, it's updated everywhere. Campaign names, product prices — same thing. Each fact lives in exactly one place.

### Advantages and Disadvantages

| Approach | Advantages | Disadvantages | Best For |
|----------|------------|---------------|----------|
| **Normalized (3NF)** | No data redundancy, data integrity (update one place), smaller storage, prevents insert/update/delete anomalies | Slow reads (many JOINs required), complex queries for reporting, not optimized for analytics | OLTP systems — applications, transactions (PostgreSQL, MySQL) |
| **Denormalized** | Fast reads (fewer or no JOINs), simpler queries for analysts, optimized for dashboards and BI tools | Data redundancy (same info in multiple places), update anomalies (change one place, miss another), larger storage | OLAP systems — warehouses, reporting (BigQuery, Redshift) |
| **Star Schema** | Simple to understand (fact + dimensions), fast analytical queries, BI tools designed for it, flexible (any dimension × any measure) | Redundancy in dimension tables, requires upfront design, fact table can grow very large, not ideal for OLTP workloads | Data warehouses — the industry standard for analytics |
| **Snowflake Schema** | Less redundancy than star (dimensions are further normalized), saves storage on very large dimension tables | More complex queries (more JOINs through sub-dimensions), slower reads than star, harder for analysts to understand | Large enterprises where dimension tables are huge (e.g., product catalog with deep category hierarchies) |

> **Normalization is for OLTP** (transactional databases). It prevents data anomalies and keeps the source of truth clean. Application databases (PostgreSQL, MySQL) are typically normalized.

### Denormalization — Optimizing for Reads

Normalized tables are great for writes but slow for reads — every query needs JOINs across multiple tables. For analytics (OLAP), we intentionally **denormalize**: combine tables back together so analysts can query without writing 5-table JOINs.

| Normalized (OLTP) | Denormalized (OLAP) |
|-------------------|---------------------|
| Many small tables, connected by foreign keys | Fewer wide tables, data duplicated for speed |
| Optimized for INSERT/UPDATE (one place to change) | Optimized for SELECT (everything in one table) |
| Requires JOINs to answer questions | Can answer questions with simple queries |
| Used by: applications, transactions | Used by: analysts, dashboards, reports |

> **The DE's job:** Extract from normalized OLTP sources → Transform → Load into denormalized OLAP structures. You go FROM normalized TO denormalized.

### Star Schema — The Standard OLAP Design

The **star schema** is the most common way to organize data in a warehouse. It has two types of tables:

| Table Type | What It Contains | Examples | Characteristics |
|-----------|-----------------|----------|----------------|
| **Fact table** | Measurable events — things that happened | Calls made, orders placed, payments processed | Large (millions/billions of rows), numeric measures, foreign keys to dimensions |
| **Dimension table** | Descriptive context — the who, what, where, when | Customers, products, campaigns, dates | Small (hundreds/thousands of rows), text attributes, slowly changing |

**Why "star"?** The fact table sits in the center with dimension tables radiating out — it looks like a star:

```
                    ┌────────────┐
                    │  dim_date  │
                    │ date, month│
                    │ quarter,   │
                    │ year, day  │
                    └─────┬──────┘
                          │
┌────────────┐     ┌──────┴──────┐     ┌────────────┐
│dim_customer│─────│ fact_calls  │─────│dim_campaign │
│ name, phone│     │             │     │ name, client│
│ city, state│     │ call_id     │     │ channel     │
└────────────┘     │ duration    │     └────────────┘
                   │ amount      │
┌────────────┐     │ disposition │     ┌────────────┐
│dim_product │─────│ upsell_flag │─────│  dim_time  │
│ name, price│     │             │     │ hour, shift│
│ category   │     └─────────────┘     └────────────┘
```

### Star Schema vs. Snowflake Schema

| | Star Schema | Snowflake Schema |
|---|---|---|
| **Dimensions** | Denormalized (one flat table per dimension) | Normalized (dimension split into sub-tables) |
| **Example** | `dim_product` has columns: product_id, name, price, category, department | `dim_product` → `dim_category` → `dim_department` (3 tables) |
| **JOINs for a query** | Fewer (fact → dimension) | More (fact → dimension → sub-dimension → ...) |
| **Query speed** | Faster (fewer JOINs) | Slower (more JOINs) |
| **Redundancy** | Some (category name repeated in every product row) | Less (category stored once in its own table) |
| **Complexity** | Simple — recommended default | More complex — use only when dimension tables are very large |
| **Industry usage** | ~90% of data warehouses | ~10% — large enterprises with huge dimension tables |

> **In this program, we use star schema.** It's the industry standard for data warehouses and what you'll encounter in the vast majority of jobs. Snowflake schema is worth knowing for interviews, but star schema is what you'll build.

### Reading the Star Schema

With this design, analysts can answer business questions with simple SQL:

| Question | Query Uses |
|----------|-----------|
| "Total revenue by campaign this quarter" | fact_calls JOIN dim_campaign JOIN dim_date → GROUP BY campaign_name, quarter |
| "Which products sell best on weekends?" | fact_calls JOIN dim_product JOIN dim_date → WHERE day_type = 'weekend' GROUP BY product_name |
| "Average call duration by customer state" | fact_calls JOIN dim_customer → GROUP BY state, AVG(duration) |
| "Hourly call volume trend" | fact_calls JOIN dim_time → GROUP BY hour |

> **The power of star schema:** One design, infinite questions. Analysts combine any dimension with any measure. No new tables needed — just different GROUP BY and WHERE clauses.

### Facts vs. Dimensions — Quick Reference

| | Fact Table | Dimension Table |
|---|---|---|
| **Contains** | Measurements, metrics, events | Descriptive attributes, context |
| **Examples of columns** | amount, duration, quantity, count | name, category, city, date, status |
| **Row count** | Millions to billions | Hundreds to thousands |
| **Changes** | New rows added constantly (append) | Changes slowly over time (SCD) |
| **Foreign keys** | Yes — points to every dimension | No — has a primary key |
| **Grain** | One row = one event (one call, one order) | One row = one entity (one customer, one product) |

## 16a. Normalization Step by Step — UNF → 1NF → 2NF → 3NF

**One-line answer:** Normalization eliminates duplicate data by splitting one messy table into several clean, related tables.

Section 16 showed you the *result* of normalization. Now let's walk through the *process* — one normal form at a time, using the simplest possible example.

### Starting Point: Unnormalized Form (UNF)

A small business tracks orders in a single spreadsheet. Some cells contain multiple values:

| order_id | date | customer | phone | city | items | prices |
|----------|------|----------|-------|------|-------|--------|
| 1001 | 2026-03-10 | Alice | 313-555-1234 | Detroit | Widget A, Widget B | 29.99, 49.99 |
| 1002 | 2026-03-10 | Bob | 248-555-5678 | Troy | Widget A | 29.99 |
| 1003 | 2026-03-11 | Alice | 313-555-1234 | Detroit | Widget C | 19.99 |

**Problems:**
- `items` and `prices` have multiple values crammed into one cell — you can't query "show me all orders for Widget A" with a simple WHERE clause
- Alice's phone and city are duplicated in rows 1 and 3 — if she moves, you update multiple rows (and might miss one)
- There's no way to store a product's price independently of an order

---

### First Normal Form (1NF) — Atomic Values

> **Rule:** Every cell contains exactly **one value**. No lists, no repeating groups.

Split the multi-value cells into separate rows:

| order_id | date | customer | phone | city | item | price |
|----------|------|----------|-------|------|------|-------|
| 1001 | 2026-03-10 | Alice | 313-555-1234 | Detroit | Widget A | 29.99 |
| 1001 | 2026-03-10 | Alice | 313-555-1234 | Detroit | Widget B | 49.99 |
| 1002 | 2026-03-10 | Bob | 248-555-5678 | Troy | Widget A | 29.99 |
| 1003 | 2026-03-11 | Alice | 313-555-1234 | Detroit | Widget C | 19.99 |

**What changed and why:**
- The `items` column (which held "Widget A, Widget B") was split: order 1001 is now **two rows** — one per item. The column was renamed `item` (singular).
- The `prices` column was split the same way: each row has exactly one price for exactly one item.
- The row count went from **3 to 4** — order 1001 expanded into 2 rows because Alice ordered 2 items.
- The primary key changed: `order_id` alone no longer uniquely identifies a row (1001 appears twice). The composite key is now **(order_id + item)** — together they uniquely identify each row.

**What's still wrong:** Alice's name, phone, and city are repeated in rows 1, 2, and 4 — three places. If she changes her phone number, you'd have to find and update all three. Miss one and your data is inconsistent.

---

### Second Normal Form (2NF) — No Partial Dependencies

> **Rule:** Every non-key column depends on the **entire** key, not just part of it.

Our composite key is **(order_id + item)**. Ask for each column: "Does this depend on both parts of the key, or just one?"

| Column | Depends on order_id? | Depends on item? | Depends on BOTH? | Problem? |
|--------|---------------------|------------------|-------------------|----------|
| `date` | Yes — the order has a date | No — doesn't matter which item | **Partial** — only needs order_id | Yes |
| `customer` | Yes — the order has a customer | No | **Partial** | Yes |
| `phone` | Yes (through customer) | No | **Partial** | Yes |
| `city` | Yes (through customer) | No | **Partial** | Yes |
| `price` | No — not about the order | Yes — Widget A is always 29.99 | **Partial** — only needs item | Yes |

Every column has a partial dependency. None of them need *both* `order_id` AND `item` together.

**Fix:** Move partially-dependent columns into their own tables — group them by which part of the key they actually depend on.

**orders** (columns that depend on order_id only):

| order_id | date | customer | phone | city |
|----------|------|----------|-------|------|
| 1001 | 2026-03-10 | Alice | 313-555-1234 | Detroit |
| 1002 | 2026-03-10 | Bob | 248-555-5678 | Troy |
| 1003 | 2026-03-11 | Alice | 313-555-1234 | Detroit |

**products** (columns that depend on item only):

| item | price |
|------|-------|
| Widget A | 29.99 |
| Widget B | 49.99 |
| Widget C | 19.99 |

**order_items** (the link — this is what's left after extraction, depends on the full key):

| order_id | item |
|----------|------|
| 1001 | Widget A |
| 1001 | Widget B |
| 1002 | Widget A |
| 1003 | Widget C |

**What changed and why:**
- We went from **1 table to 3 tables**: `orders` (order-level data), `products` (item-level data), `order_items` (the many-to-many link between them).
- `price` moved to `products` — Widget A's price (29.99) is now stored **once**, not in every order that includes it. If the price changes, update one row.
- `date`, `customer`, `phone`, `city` moved to `orders` — stored once per order, not once per item in the order.
- The original "1001, Widget A, 29.99" and "1001, Widget B, 49.99" no longer carry redundant order info (date, customer, phone, city) — that info lives in the `orders` table and is looked up via `order_id`.

**What's still wrong:** In the `orders` table, look at rows 1001 and 1003 — Alice's phone (313-555-1234) and city (Detroit) are repeated. Why? Because `phone` and `city` don't really depend on `order_id` — they depend on `customer`. That's a transitive dependency.

---

### Third Normal Form (3NF) — No Transitive Dependencies

> **Rule:** Non-key columns depend **only on the key**, not on other non-key columns.

In the `orders` table, trace the dependencies:
- `order_id` → `customer` (each order has a customer) ✓
- `customer` → `phone`, `city` (each customer has a phone and city) ← **this is the problem**

The chain is: `order_id` → `customer` → `phone`, `city`. The phone and city depend on `customer`, which depends on `order_id`. That indirect chain is a **transitive dependency** — `phone` doesn't depend *directly* on `order_id`, it depends on it *through* `customer`.

**Fix:** Extract customer details into their own table. Replace the customer name in `orders` with a `customer_id` foreign key.

**customers** (new table — customer data stored once):

| customer_id | name | phone | city |
|-------------|------|-------|------|
| 1 | Alice | 313-555-1234 | Detroit |
| 2 | Bob | 248-555-5678 | Troy |

**orders** (now references customer_id instead of storing customer details):

| order_id | date | customer_id |
|----------|------|-------------|
| 1001 | 2026-03-10 | 1 |
| 1002 | 2026-03-10 | 2 |
| 1003 | 2026-03-11 | 1 |

**products** (add a proper surrogate key):

| product_id | name | price |
|------------|------|-------|
| 100 | Widget A | 29.99 |
| 101 | Widget B | 49.99 |
| 102 | Widget C | 19.99 |

**order_items** (now uses proper foreign keys):

| order_id | product_id |
|----------|------------|
| 1001 | 100 |
| 1001 | 101 |
| 1002 | 100 |
| 1003 | 102 |

**What changed and why:**
- We went from **3 tables to 4 tables**: the `orders` table was split further — customer details (`name`, `phone`, `city`) moved into a new `customers` table.
- Alice now exists **once** — in the `customers` table. Orders 1001 and 1003 both point to `customer_id = 1` instead of repeating "Alice, 313-555-1234, Detroit" in each row.
- If Alice moves from Detroit to Ann Arbor, you update **one row** in `customers`. Every order automatically reflects her new city because they link by `customer_id`.
- We also added surrogate keys: `customer_id` (1, 2) and `product_id` (100, 101, 102) — integer IDs that replace using names as identifiers. This is standard practice.
- The `orders` table is now just 3 columns: `order_id`, `date`, `customer_id`. Lean and clean.

**Result:** Every fact is stored exactly once. Alice's phone number → one place. Widget A's price → one place. Change it once, it's correct everywhere. No more update anomalies.

---

### The Journey at a Glance

| Step | Tables | What Moved | Why |
|------|--------|-----------|-----|
| **UNF** | 1 table | Everything in one spreadsheet | Starting point — multi-value cells |
| **→ 1NF** | 1 table (more rows) | Multi-value cells split into separate rows | Each cell must hold exactly one value |
| **→ 2NF** | 3 tables | `price` → `products`, order info → `orders`, link → `order_items` | Columns that depend on only *part* of the key get their own table |
| **→ 3NF** | 4 tables | `customer`, `phone`, `city` → `customers` | Columns that depend on another non-key column (not directly on the key) get their own table |

### One Sentence Each

| Form | One-Liner |
|------|-----------|
| **1NF** | Every cell has one value (no lists) |
| **2NF** | Every column depends on the *whole* key (no partial dependencies) |
| **3NF** | Every column depends on *nothing but* the key (no transitive dependencies) |

> **The classic mnemonic:** "The key, the whole key, and nothing but the key — so help me Codd." (Edgar F. Codd invented relational databases and normal forms in 1970.)

### How 3NF Connects to Star Schema — Are They the Same?

**No. They solve different problems for different systems.**

Normalization (3NF) is how you organize data in **OLTP databases** — the systems that run applications (PostgreSQL, MySQL, SQL Server). The goal is data integrity: no redundancy, no update anomalies.

Star schema is how you organize data in **OLAP databases** — the systems that run analytics (BigQuery, Redshift, Snowflake). The goal is query speed: analysts can answer questions without writing complex multi-table JOINs.

| | Normalization (3NF) | Star Schema |
|---|---|---|
| **Purpose** | Prevent data anomalies (duplicates, inconsistencies) | Fast analytical queries |
| **Used in** | OLTP — applications, transactions | OLAP — warehouses, dashboards |
| **Design priority** | Write efficiency (INSERT, UPDATE, DELETE) | Read efficiency (SELECT, GROUP BY, JOIN) |
| **Redundancy** | Zero — each fact stored once | Intentional — dimensions are denormalized for simplicity |
| **Number of tables** | Many small, interconnected tables | Few: 1 fact table + several dimension tables |
| **JOINs required** | Many (data is split across many tables) | Few (fact → dimension, one hop) |
| **Example** | `orders` → `order_items` → `products` → `categories` | `fact_orders` JOIN `dim_product` (product + category in one table) |

**But they share the same vocabulary:**

The `customers` table in 3NF and the `dim_customer` table in a star schema look very similar — both have customer_id, name, phone, city. The difference is intent:

| 3NF Table | Star Schema Equivalent | What's Different |
|-----------|----------------------|-----------------|
| `customers` | `dim_customer` | Star schema adds SCD Type 2 columns (effective_date, end_date, is_current) to track history |
| `products` | `dim_product` | Star schema may flatten category/department into the same table (denormalized) |
| `orders` + `order_items` | `fact_calls` or `fact_orders` | Star schema pre-joins and adds measures (duration, amount) — one row per event |

**The DE's job is the bridge:**

```
[OLTP Source]           [DE Pipeline]           [OLAP Warehouse]
                        
 customers (3NF)  ───→  Transform  ───→  dim_customer (star)
 products (3NF)   ───→  + Enrich   ───→  dim_product (star)  
 orders (3NF)     ───→  + Clean    ───→  fact_orders (star)
 order_items (3NF)───→  + Join     ───→  (merged into fact)
```

You extract FROM normalized 3NF sources, transform the data, and load INTO a denormalized star schema. Understanding both sides is what makes you effective.

> **Interview phrasing:** "Normalization eliminates redundancy in transactional databases. Star schema intentionally reintroduces some redundancy to optimize analytical queries. As a data engineer, I extract from normalized sources and load into star schemas — the pipeline is the bridge between the two."
