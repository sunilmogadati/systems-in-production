```python
# === PYTHON LANGUAGE RULES ===
# WHY: These rules are where Python differs from Java, C#, and
# JavaScript. Know these and you will never be confused reading Python code.

# --- 1. CASE SENSITIVITY ---
# Python is case-sensitive. 'Model', 'model', and 'MODEL' are three different variables.
Model = "random_forest"
model = "logistic_regression"
MODEL = "svm"
print(f"Model={Model}, model={model}, MODEL={MODEL}")

# --- 2. INDENTATION IS SYNTAX ---
# No braces. Indent = block. Standard: 4 spaces (never tabs).
x = 10
if x > 5:
    print("  inside the if block (indented 4 spaces)")
    if x > 8:
        print("  nested block (indented 8 spaces)")
print("outside the if block (no indent)")
# Wrong indentation = IndentationError (your code will not run)

# --- 3. LINE CONTINUATION ---
# Long lines can be split with a backslash (\)
total = 1 + 2 + 3 + \
        4 + 5 + 6
print(f"Line continuation: {total}")
# Inside parentheses, brackets, or braces, no backslash needed:
features = [
    "age", "income",
    "credit_score", "tenure"
]
print(f"Multi-line list: {features}")

# --- 4. SEMICOLONS ---
# Semicolons CAN separate multiple statements on one line.
# Legal but discouraged. Pythonic style: one statement per line.
a = 1; b = 2; c = 3
print(f"Semicolons work: a={a}, b={b}, c={c}")

# --- 5. COMMENTS ---
# Single-line: use #
# Multi-line: no block comment syntax. Use # on each line.
# Triple-quoted strings (''' or \"\"\") are docstrings, not block comments.
# In a notebook cell, a bare triple-quoted string creates an unused string object.

# --- 6. VARIABLE NAMING CONVENTIONS ---
# snake_case for variables and functions (not camelCase like Java)
user_count = 100           # good
# userCount = 100          # legal but not Pythonic
MAX_RETRIES = 5            # UPPER_CASE for constants
_internal_cache = {}       # leading underscore = private by convention
# Names must start with a letter or underscore, never a digit.
# 2nd_attempt = "bad"     # SyntaxError
second_attempt = "good"
print(f"Naming: user_count={user_count}, MAX_RETRIES={MAX_RETRIES}")

# You Should See:
# Model=random_forest, model=logistic_regression, MODEL=svm
# inside the if block (indented 4 spaces)
# nested block (indented 8 spaces)
# outside the if block (no indent)
# Line continuation: 21
# Multi-line list: ['age', 'income', 'credit_score', 'tenure']
# Semicolons work: a=1, b=2, c=3
# Naming: user_count=100, MAX_RETRIES=5
```

```python
# === DYNAMIC TYPING, TYPE ERRORS, AND USER INPUT ===
# WHY: Python is dynamically typed (type assigned at runtime, not in code).
# It is also strongly typed (no implicit conversions like JavaScript).

# --- DYNAMIC TYPING ---
# No type declarations. The type comes from the value you assign.
x = 42              # x is an int
print(f"x = {x}, type = {type(x).__name__}")
x = "now a string"  # same variable, now a str — Python allows this
print(f"x = {x}, type = {type(x).__name__}")
x = [1, 2, 3]       # now a list
print(f"x = {x}, type = {type(x).__name__}")

# --- CAN YOU DECLARE WITHOUT ASSIGNING? ---
# No. In Python, a variable only exists after assignment.
# There is no equivalent of Java's 'int x;' without a value.
# Closest pattern: assign None as a placeholder.
result = None        # exists, but has no meaningful value yet
print(f"Placeholder: result = {result}")

# --- NameError: USING A VARIABLE BEFORE DECLARING IT ---
try:
    print(undefined_variable)
except NameError as e:
    print(f"NameError: {e}")

# --- STRONG TYPING: str + int FAILS ---
# JavaScript: "5" + 3 = "53" (silent conversion)
# Python: "5" + 3 = TypeError (refuses to guess)
try:
    result = "5" + 3
except TypeError as e:
    print(f"TypeError: {e}")

# Fix: explicitly convert
print(f"Correct: {'5'} + {3} = {int('5') + 3}")  # convert str to int
print(f"Or:      {'5'} + {3} = {'5' + str(3)}")   # convert int to str

# --- ILLEGAL TYPE CONVERSION ---
# What happens if you convert text to a number?
try:
    bad = int("hello")
except ValueError as e:
    print(f"ValueError: {e}")

# Valid conversions:
print(f"float('3.14') = {float('3.14')}")   # str -> float
print(f"int(3.99)     = {int(3.99)}")        # float -> int (truncates, no rounding)
print(f"str(42)       = '{str(42)}'")         # int -> str

# --- USER INPUT ---
# input() prompts the user and returns a string.
# Like Java's Scanner or C's scanf, but always returns str.
# Uncomment to try interactively:
# name = input("Enter your name: ")
# print(f"Hello, {name}!")
# age = int(input("Enter your age: "))  # must convert to int
# print(f"Next year you will be {age + 1}")
print("(input() is commented out — uncomment to try interactively)")

# You Should See:
# x = 42, type = int
# x = now a string, type = str
# x = [1, 2, 3], type = list
# Placeholder: result = None
# NameError: name 'undefined_variable' is not defined
# TypeError: can only concatenate str (not "int") to str
# Correct: 5 + 3 = 8
# Or:      5 + 3 = 53
# ValueError: invalid literal for int() with base 10: 'hello'
# float('3.14') = 3.14
# int(3.99)     = 3
# str(42)       = '42'
```

```python
# === DICTIONARIES -- KEY-VALUE PAIRS ===
# WHY: Dicts map names to values. Model configs, API responses,
# feature stores -- all are dictionaries.

# Create a dict
config = {
    "model": "random_forest",
    "n_estimators": 100,
    "max_depth": 10,
    "random_state": 42
}
print(f"type: {type(config).__name__}")
print(f"Model: {config['model']}")

# .get() with default -- safe access (no KeyError)
lr = config.get('learning_rate', 'not set')
print(f"Learning rate: {lr}")

# Add, update, delete
config["min_samples_split"] = 5    # add new key
config["max_depth"] = 8            # update existing
del config["random_state"]          # delete
print(f"After mutations: {config}")

# Iterate keys, values, key-value pairs
print("\nAll keys:", list(config.keys()))
print("All values:", list(config.values()))
print("All items:")
for key, value in config.items():
    print(f"  {key}: {value}")

# {} = empty dict (not set!). Use set() for empty set.
empty_dict = {}
empty_set = set()
print(f"\n{{}} is {type(empty_dict).__name__}, set() is {type(empty_set).__name__}")

# Duplicate keys -- last value wins silently
dupes = {"a": 1, "b": 2, "a": 99}
print(f"Duplicate keys: {dupes}")   # {'a': 99, 'b': 2}

# Merge dicts (Python 3.9+)
defaults = {"lr": 0.01, "epochs": 10}
overrides = {"lr": 0.001}
merged = defaults | overrides
print(f"Merged: {merged}")

# Dict comprehension
features = ["age", "income", "credit_score", "tenure"]
feature_index = {name: i for i, name in enumerate(features)}
print(f"Feature index: {feature_index}")

# You Should See:
# type: dict
# Model: random_forest
# Learning rate: not set
# {} is dict, set() is set
# Duplicate keys: {'a': 99, 'b': 2}
# Merged: {'lr': 0.001, 'epochs': 10}
```

```python
# === TUPLES -- IMMUTABLE LISTS ===
# WHY: Tuples are ordered collections that CANNOT be changed.
# Use them for: function return values, dict keys, fixed configs.

# Create a tuple
point = (10, 20, 30)
print(f"point: {point}, type: {type(point).__name__}")

# Indexing and slicing work the same as lists
print(f"point[0]:   {point[0]}")
print(f"point[-1]:  {point[-1]}")
print(f"point[:2]:  {point[:2]}")

# But you CANNOT modify (this is the key difference from lists)
try:
    point[0] = 99
except TypeError as e:
    print(f"TypeError: {e}")

# Unpacking -- the #1 reason you will use tuples
x, y, z = point
print(f"Unpacked: x={x}, y={y}, z={z}")

# Functions return tuples for multiple values
def min_max(nums):
    return min(nums), max(nums)

lo, hi = min_max([3, 1, 4, 1, 5, 9])
print(f"min={lo}, max={hi}")

# Star unpacking -- collect the rest
first, *middle, last = (1, 2, 3, 4, 5)
print(f"first={first}, middle={middle}, last={last}")

# Convert between list and tuple
as_list = list(point)         # tuple -> list
as_tuple = tuple([4, 5, 6])   # list -> tuple
print(f"list(point): {as_list}, tuple([4,5,6]): {as_tuple}")

# You Should See:
# point: (10, 20, 30), type: tuple
# point[0]: 10
# TypeError: 'tuple' object does not support item assignment
# Unpacked: x=10, y=20, z=30
# min=1, max=9
# first=1, middle=[2, 3, 4], last=5
```

```python
# === SETS -- UNIQUE, UNORDERED COLLECTIONS ===
# WHY: Sets deduplicate data and test membership in O(1).
# In ML, use sets to find overlapping features, missing IDs, duplicates.

# Create a set — duplicates auto-removed
tags = {"python", "ai", "ml", "python"}
print(f"tags: {tags}")            # only 3 items
print(f"type: {type(tags).__name__}")

# set() from a list (common dedup pattern)
ids = ["C001", "C002", "C003", "C002", "C001"]
unique = set(ids)
print(f"\nRaw: {len(ids)}, Unique: {len(unique)}, Dupes: {len(ids) - len(unique)}")

# Membership test — O(1), much faster than list
print(f"'python' in tags: {'python' in tags}")

# Set operations — compare datasets
train_features = {"age", "income", "credit_score", "tenure"}
test_features = {"age", "income", "spend", "region"}
print(f"\nIn both:    {train_features & test_features}")
print(f"Train only: {train_features - test_features}")
print(f"All:        {train_features | test_features}")

# Modify: add, remove, discard
tags.add("deep_learning")
tags.discard("ml")        # safe — no error if missing
print(f"Modified: {tags}")

# Convert: set -> list, list -> set, set -> tuple
as_list = list(unique)
as_tuple = tuple(unique)
print(f"set -> list: {as_list}")

# You Should See:
# tags: {'python', 'ai', 'ml'}
# Raw: 5, Unique: 3, Dupes: 2
# 'python' in tags: True
# In both: {'age', 'income'}
```

```python
# === LOOPS AND CONDITIONALS ===
# WHY: Iterate over data, classify results, filter batches.

predictions = [0.92, 0.45, 0.88, 0.31, 0.97, 0.62]
threshold = 0.5

# for loop with enumerate (gives index + value)
for i, score in enumerate(predictions):
    label = "positive" if score >= threshold else "negative"
    print(f"  [{i}] {score:.2f} -> {label}")

# Count positives
positive_count = sum(1 for s in predictions if s >= threshold)
print(f"\nPositive: {positive_count}/{len(predictions)}")

# You Should See:
#   [0] 0.92 -> positive
#   [1] 0.45 -> negative
#   [2] 0.88 -> positive
#   [3] 0.31 -> negative
#   [4] 0.97 -> positive
#   [5] 0.62 -> positive
# Positive: 4/6
```

## Section 2: Functions and Data Flow

Functions are how you organize code for reuse. In AI work, you will write functions to preprocess data, compute metrics, and build pipelines.

**Deep dive:** [![Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/sunilmogadati/systems-in-production/blob/main/implementation/notebooks/Python_Functions_Classes.ipynb) Functions and Classes notebook

```python
# === FUNCTIONS ===
# WHY: Reusable logic. compute_metrics() will be called hundreds
# of times across training runs and evaluations.

def compute_metrics(y_true, y_pred, threshold=0.5):
    """Compute accuracy and positive rate from predictions."""
    labels = [1 if p >= threshold else 0 for p in y_pred]
    correct = sum(1 for t, l in zip(y_true, labels) if t == l)
    accuracy = correct / len(y_true)
    positive_rate = sum(labels) / len(labels)
    return {"accuracy": accuracy, "positive_rate": positive_rate}

# Test it
actual =      [1, 0, 1, 0, 1, 1]
predicted =   [0.92, 0.45, 0.88, 0.31, 0.97, 0.62]

result = compute_metrics(actual, predicted)
print(f"Accuracy:      {result['accuracy']:.1%}")
print(f"Positive rate: {result['positive_rate']:.1%}")

# With a different threshold
strict = compute_metrics(actual, predicted, threshold=0.9)
print(f"\nStrict (0.9):  {strict['accuracy']:.1%}")

# You Should See:
# Accuracy:      100.0%
# Positive rate: 66.7%
# Strict (0.9):  66.7%
```

```python
# === LIST COMPREHENSIONS ===
# WHY: The Pythonic way to transform data. Replaces map/filter
# patterns from Java streams or JavaScript array methods.

scores = [0.92, 0.45, 0.88, 0.31, 0.97, 0.62]

# Transform: scale to percentage
percentages = [round(s * 100, 1) for s in scores]
print(f"Percentages: {percentages}")

# Filter: keep only high confidence
high_conf = [s for s in scores if s >= 0.8]
print(f"High confidence: {high_conf}")

# Transform + filter: labels for high confidence only
labels = [f"PASS ({s:.0%})" for s in scores if s >= 0.8]
print(f"Labels: {labels}")

# Dict comprehension: feature name -> index mapping
features = ["age", "income", "credit_score", "tenure"]
feature_index = {name: i for i, name in enumerate(features)}
print(f"Feature index: {feature_index}")

# You Should See:
# Percentages: [92.0, 45.0, 88.0, 31.0, 97.0, 62.0]
# High confidence: [0.92, 0.88, 0.97]
# Labels: ['PASS (92%)', 'PASS (88%)', 'PASS (97%)']
# Feature index: {'age': 0, 'income': 1, 'credit_score': 2, 'tenure': 3}
```

```python
# === LAMBDA FUNCTIONS ===
# WHY: Short throwaway functions used in sorting, filtering,
# and Pandas .apply(). Java equivalent: x -> x * 2

# Sort models by accuracy (descending)
models = [
    {"name": "logistic", "accuracy": 0.85},
    {"name": "random_forest", "accuracy": 0.92},
    {"name": "svm", "accuracy": 0.88},
]

ranked = sorted(models, key=lambda m: m["accuracy"], reverse=True)
for rank, m in enumerate(ranked, 1):
    print(f"  {rank}. {m['name']:15s} {m['accuracy']:.1%}")

# You Should See:
#   1. random_forest   92.0%
#   2. svm             88.0%
#   3. logistic         85.0%
```

## Section 3: NumPy -- Arrays for AI

NumPy (Numerical Python) arrays are the foundation of all AI/ML (Artificial Intelligence / Machine Learning) in Python. Every framework -- scikit-learn, PyTorch, TensorFlow -- converts data to NumPy arrays internally.

**Deep dive:** [![Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/sunilmogadati/systems-in-production/blob/main/implementation/notebooks/Python_NumPy_Pandas.ipynb) NumPy and Pandas notebook

```python
# === CREATING ARRAYS ===
# WHY: Arrays have a shape and a dtype. Understanding these
# prevents the most common ML error: "expected shape (N, 4), got (N,)"

import numpy as np

# 1D array -- a vector of predictions
scores = np.array([0.92, 0.45, 0.88, 0.31, 0.97])
print(f"scores: {scores}")
print(f"shape:  {scores.shape}  dtype: {scores.dtype}")

# 2D array -- a feature matrix (5 customers, 3 features each)
# Columns: age, income, credit_score
customers = np.array([
    [25, 50000, 720],
    [35, 80000, 680],
    [45, 120000, 750],
    [28, 45000, 640],
    [52, 95000, 710]
])
print(f"\ncustomers shape: {customers.shape}")  # (5, 3) = 5 rows, 3 columns

# You Should See:
# scores: [0.92 0.45 0.88 0.31 0.97]
# shape:  (5,)  dtype: float64
# customers shape: (5, 3)
```

```python
# === VECTORIZED MATH ===
# WHY: NumPy operates on entire arrays at once -- 10-100x faster
# than Python loops. This is how ML frameworks process data.

import numpy as np

customers = np.array([
    [25, 50000, 720],
    [35, 80000, 680],
    [45, 120000, 750],
    [28, 45000, 640],
    [52, 95000, 710]
])

# Column statistics -- axis=0 means "down each column"
print(f"Mean age:    {customers[:, 0].mean():.1f}")
print(f"Max income:  {customers[:, 1].max():,}")
print(f"Min credit:  {customers[:, 2].min()}")

# Feature scaling (standardization) -- critical for ML
# Formula: z = (x - mean) / std
means = customers.mean(axis=0)
stds = customers.std(axis=0)
scaled = (customers - means) / stds   # broadcasts automatically

print(f"\nBefore scaling (row 0): {customers[0]}")
print(f"After scaling (row 0):  {scaled[0].round(3)}")
print(f"Scaled means: {scaled.mean(axis=0).round(10)}")  # ~0
print(f"Scaled stds:  {scaled.std(axis=0).round(10)}")   # ~1

# You Should See:
# Mean age:    37.0
# Max income:  120000
# Min credit:  640
# Before scaling (row 0): [   25 50000   720]
# After scaling (row 0):  [-1.177 -1.036  0.535]
# Scaled means: [0. 0. 0.]
# Scaled stds:  [1. 1. 1.]
```

```python
# === BOOLEAN INDEXING AND DOT PRODUCTS ===
# WHY: Boolean indexing filters data without loops.
# Dot products are the core operation in neural networks: y = X @ w + b

import numpy as np

scores = np.array([0.92, 0.45, 0.88, 0.31, 0.97])

# Boolean mask -- filter without loops
high = scores[scores >= 0.8]
print(f"Scores >= 0.8: {high}")
print(f"Count:         {(scores >= 0.8).sum()}")

# Dot product -- the math behind linear regression
# Prediction = features @ weights + bias
features = np.array([25, 50000, 720])     # one customer
weights = np.array([0.01, 0.00001, 0.005])  # learned weights
bias = -2.0

prediction = features @ weights + bias
print(f"\nFeatures: {features}")
print(f"Weights:  {weights}")
print(f"Prediction: {prediction:.4f}")

# You Should See:
# Scores >= 0.8: [0.92 0.88 0.97]
# Count:         3
# Features: [   25 50000   720]
# Weights:  [1.e-02 1.e-05 5.e-03]
# Prediction: 1.8500
```

```python
# === FILTER AND SELECT ===
# WHY: "Show me all churned customers" -- this is the query
# you write 100 times a day.

# Filter rows
churned = customers[customers["churned"] == True]
print(f"Churned customers: {len(churned)}")
print(churned[["customer_id", "tenure_months", "monthly_spend"]].to_string(index=False))

# Multiple conditions: high income AND long tenure
loyal_high = customers[(customers["income"] > 70000) & (customers["tenure_months"] > 12)]
print(f"\nHigh income + long tenure: {len(loyal_high)}")

# .query() -- SQL-like syntax (cleaner for complex filters)
result = customers.query("age < 30 and churned == True")
print(f"\nYoung churned customers:")
print(result[["customer_id", "age", "tenure_months"]].to_string(index=False))

# You Should See:
# Churned customers: 3
# customer_id  tenure_months  monthly_spend
#        C004              3          29.99
#        C006              6          49.99
#        C008              2          19.99
```

```python
# === DATA CLEANING ===
# WHY: Real data has nulls, duplicates, and quality issues.
# You MUST clean before modeling -- garbage in, garbage out.

# Check for nulls
print("Null counts:")
print(customers.isnull().sum()[customers.isnull().sum() > 0])

# Fill nulls: credit_score with median, monthly_spend with median
customers["credit_score"] = customers["credit_score"].fillna(
    customers["credit_score"].median()
)
customers["monthly_spend"] = customers["monthly_spend"].fillna(
    customers["monthly_spend"].median()
)

# Verify: no more nulls
print(f"\nNulls remaining: {customers.isnull().sum().sum()}")

# You Should See:
# Null counts:
# credit_score     1
# monthly_spend    1
# Nulls remaining: 0
```

```python
# === GROUPBY -- THE SQL GROUP BY EQUIVALENT ===
# WHY: "Average spend by churn status" is the kind of question
# that tells you whether a feature is predictive.

by_churn = customers.groupby("churned").agg(
    count=("customer_id", "count"),
    avg_tenure=("tenure_months", "mean"),
    avg_spend=("monthly_spend", "mean"),
    avg_credit=("credit_score", "mean")
).round(1)

print("Metrics by churn status:")
print(by_churn)

# You Should See:
# Churned customers have lower tenure, lower spend, lower credit.
# This tells us these features are likely predictive of churn.
```

## Section 6: Your First ML Model

Now we bring everything together. We will train a Random Forest classifier to predict customer churn -- using the data we just explored and cleaned.

**The ML workflow:**
1. Select features (X) and target (y)
2. Split into training and test sets
3. Train the model
4. Evaluate on unseen data
5. Interpret what the model learned

**Deep dive:** [![Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/sunilmogadati/systems-in-production/blob/main/implementation/notebooks/Linear_Regression.ipynb) Linear Regression notebook | [![Colab](https://colab.research.google.com/assets/colab-badge.svg)](https://colab.research.google.com/github/sunilmogadati/systems-in-production/blob/main/implementation/notebooks/ML_Fundamentals.ipynb) ML Fundamentals notebook

```python
# === STEP 1: PREPARE FEATURES AND TARGET ===
# WHY: ML models need numeric arrays. X = features (input),
# y = target (what we predict). We drop non-numeric columns.

import pandas as pd
import numpy as np

# Use the customer data we already cleaned
feature_cols = ["age", "income", "credit_score", "tenure_months", "monthly_spend"]

X = customers[feature_cols]       # feature matrix
y = customers["churned"].astype(int)  # target: 0 = retained, 1 = churned

print(f"Features (X): {X.shape}")
print(f"Target (y):   {y.shape}")
print(f"Churn rate:   {y.mean():.1%}")
print(f"\nFeature columns: {feature_cols}")

# You Should See:
# Features (X): (10, 5)
# Target (y):   (10,)
# Churn rate:   30.0%
```

```python
# === STEP 2: TRAIN/TEST SPLIT ===
# WHY: We train on one portion and evaluate on another.
# If we test on training data, accuracy is artificially high.
# This is called "data leakage" -- a common ML mistake.

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.3,       # 30% for testing
    random_state=42,     # reproducible split
    stratify=y           # keep churn ratio in both sets
)

print(f"Training: {X_train.shape[0]} samples")
print(f"Testing:  {X_test.shape[0]} samples")
print(f"Train churn rate: {y_train.mean():.1%}")
print(f"Test churn rate:  {y_test.mean():.1%}")

# You Should See:
# Training: 7 samples
# Testing:  3 samples
# Both churn rates should be close to 30%
```

```python
# === STEP 3: TRAIN THE MODEL ===
# WHY: Random Forest builds many decision trees and votes.
# It works well out of the box -- a great first model.

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(
    n_estimators=100,    # 100 decision trees
    max_depth=3,         # limit depth to prevent overfitting on small data
    random_state=42
)

model.fit(X_train, y_train)
print("Model trained!")
print(f"Number of trees: {model.n_estimators}")
print(f"Features used:   {model.n_features_in_}")

# You Should See:
# Model trained!
# Number of trees: 100
# Features used:   5
```

```python
# === STEP 4: EVALUATE ===
# WHY: Accuracy alone is misleading. A model that always says
# "not churned" gets 70% accuracy on our data -- but is useless.
# We need precision, recall, and a confusion matrix.

from sklearn.metrics import accuracy_score, classification_report, confusion_matrix

# Predict on test set
y_pred = model.predict(X_test)

print(f"Accuracy: {accuracy_score(y_test, y_pred):.1%}")
print(f"\nClassification Report:")
print(classification_report(y_test, y_pred, target_names=["Retained", "Churned"]))

print("Confusion Matrix:")
cm = confusion_matrix(y_test, y_pred)
print(f"  Predicted:  Retained  Churned")
print(f"  Retained:   {cm[0][0]:>5}     {cm[0][1]:>5}")
print(f"  Churned:    {cm[1][0]:>5}     {cm[1][1]:>5}")

# You Should See:
# Accuracy, precision, recall, and f1-score for both classes.
# The confusion matrix shows true positives, false positives, etc.
# Note: With only 10 samples, results vary -- this is for learning the workflow.
```

```python
# === STEP 5: FEATURE IMPORTANCE ===
# WHY: Understanding WHICH features matter is more valuable than
# the prediction itself. This is what you present to stakeholders.

import matplotlib.pyplot as plt
import numpy as np

importances = model.feature_importances_
sorted_idx = np.argsort(importances)

plt.figure(figsize=(8, 4))
plt.barh(range(len(sorted_idx)), importances[sorted_idx], color="steelblue")
plt.yticks(range(len(sorted_idx)), [feature_cols[i] for i in sorted_idx])
plt.xlabel("Feature Importance")
plt.title("What Predicts Churn?")
plt.tight_layout()
plt.show()

# Print ranked importance
print("\nFeature importance (ranked):")
for i in sorted_idx[::-1]:
    print(f"  {feature_cols[i]:20s} {importances[i]:.3f}")

# You Should See:
# A horizontal bar chart showing which features the model relies on most.
# Tenure and monthly_spend should rank high (short tenure + low spend = churn).
```

```python
# === STEP 6: MAKE A PREDICTION ON NEW DATA ===
# WHY: This is the whole point -- predict churn for a new customer
# who was not in the training data.

import pandas as pd

# A new customer walks in
new_customer = pd.DataFrame([{
    "age": 26,
    "income": 42000,
    "credit_score": 630,
    "tenure_months": 2,
    "monthly_spend": 24.99
}])

# Predict
prediction = model.predict(new_customer)[0]
probability = model.predict_proba(new_customer)[0]

print("New Customer:")
print(new_customer.to_string(index=False))
print(f"\nPrediction:  {'WILL CHURN' if prediction == 1 else 'WILL STAY'}")
print(f"Confidence:  {max(probability):.1%}")
print(f"Probabilities: Retain={probability[0]:.1%}, Churn={probability[1]:.1%}")

# You Should See:
# Young, low-income, low-tenure customer -> likely to churn.
# This is the pattern our model learned from the training data.
```
